```python
# Simple To-Do List

tasks = []

while True:
    print("\n1. Add Task")
    print("2. View Tasks")
    print("3. Remove Task")
    print("4. Exit")

    choice = input("Choose an option: ")

    if choice == "1":
        task = input("Enter task: ")
        tasks.append(task)
        print("Task added!")

    elif choice == "2":
        print("\nYour Tasks:")
        for i, task in enumerate(tasks, start=1):
            print(f"{i}. {task}")

    elif choice == "3":
        task_num = int(input("Enter task number to remove: "))
        if 0 < task_num <= len(tasks):
            removed = tasks.pop(task_num - 1)
            print(f"Removed: {removed}")
        else:
            print("Invalid task number!")

    elif choice == "4":
        print("Goodbye!")
        break

    else:
        print("Invalid choice!")
```

```python
# Prime Number Checker

n = int(input("Enter a number: "))

if n <= 1:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            is_prime = False
            break

    if is_prime:
        print("Prime Number")
    else:
        print("Not Prime")
```

```python
# Mini Blockchain in Python

import hashlib
import time

class Block:
    def __init__(self, index, data, previous_hash):
        self.index = index
        self.timestamp = time.time()
        self.data = data
        self.previous_hash = previous_hash
        self.nonce = 0
        self.hash = self.mine_block()

    def calculate_hash(self):
        text = str(self.index) + str(self.timestamp) + \
               str(self.data) + str(self.previous_hash) + \
               str(self.nonce)

        return hashlib.sha256(text.encode()).hexdigest()

    def mine_block(self, difficulty=4):
        target = "0" * difficulty

        while True:
            self.hash = self.calculate_hash()

            if self.hash.startswith(target):
                return self.hash

            self.nonce += 1


class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        return Block(0, "Genesis Block", "0")

    def get_latest_block(self):
        return self.chain[-1]

    def add_block(self, data):
        previous = self.get_latest_block()

        new_block = Block(
            len(self.chain),
            data,
            previous.hash
        )

        self.chain.append(new_block)

    def is_valid(self):
        for i in range(1, len(self.chain)):
            current = self.chain[i]
            previous = self.chain[i - 1]

            if current.hash != current.calculate_hash():
                return False

            if current.previous_hash != previous.hash:
                return False

        return True


# Run Blockchain
coin = Blockchain()

coin.add_block("Alice sends 5 BTC to Bob")
coin.add_block("Bob sends 2 BTC to Charlie")
coin.add_block("Charlie sends 1 BTC to David")

for block in coin.chain:
    print("\nBLOCK:", block.index)
    print("Data:", block.data)
    print("Hash:", block.hash)
    print("Previous:", block.previous_hash)

print("\nBlockchain Valid:", coin.is_valid())
```

```python
# Mini Banking System using OOP

class BankAccount:
    bank_name = "Python Bank"

    def __init__(self, owner, balance=0):
        self.owner = owner
        self.__balance = balance   # private variable

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}")

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrawn ${amount}")
        else:
            print("Insufficient Balance!")

    def check_balance(self):
        print(f"Current Balance: ${self.__balance}")

    @classmethod
    def bank_info(cls):
        print(f"Welcome to {cls.bank_name}")

    @staticmethod
    def calculate_interest(balance, rate):
        return (balance * rate) / 100


# Inheritance
class SavingsAccount(BankAccount):
    def __init__(self, owner, balance=0, limit=500):
        super().__init__(owner, balance)
        self.limit = limit

    def withdraw(self, amount):
        if amount > self.limit:
            print("Withdrawal limit exceeded!")
        else:
            super().withdraw(amount)


# Main Program
BankAccount.bank_info()

user1 = SavingsAccount("Alice", 1000)

user1.deposit(500)
user1.withdraw(200)
user1.withdraw(700)

user1.check_balance()

interest = BankAccount.calculate_interest(1000, 5)
print("Interest:", interest)
```

```python
import sqlite3
import threading
import logging
from datetime import datetime
from functools import wraps

# ---------------- LOGGING ----------------
logging.basicConfig(
    filename="task_manager.log",
    level=logging.INFO,
    format="%(asctime)s - %(message)s"
)

# ---------------- DECORATOR ----------------
def log_action(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        logging.info(f"Action: {func.__name__} called")
        return func(*args, **kwargs)
    return wrapper


# ---------------- DATABASE LAYER ----------------
class Database:
    def __init__(self, db_name="tasks.db"):
        self.conn = sqlite3.connect(db_name, check_same_thread=False)
        self.cursor = self.conn.cursor()
        self.create_table()

    def create_table(self):
        self.cursor.execute("""
            CREATE TABLE IF NOT EXISTS tasks (
                id INTEGER PRIMARY KEY AUTOINCREMENT,
                title TEXT NOT NULL,
                status TEXT NOT NULL,
                created_at TEXT NOT NULL
            )
        """)
        self.conn.commit()

    def execute(self, query, params=()):
        self.cursor.execute(query, params)
        self.conn.commit()
        return self.cursor


# ---------------- TASK MODEL ----------------
class Task:
    def __init__(self, title, status="pending"):
        self.title = title
        self.status = status
        self.created_at = datetime.now().strftime("%Y-%m-%d %H:%M:%S")


# ---------------- SERVICE LAYER ----------------
class TaskService:
    def __init__(self, db: Database):
        self.db = db

    @log_action
    def add_task(self, task: Task):
        self.db.execute(
            "INSERT INTO tasks (title, status, created_at) VALUES (?, ?, ?)",
            (task.title, task.status, task.created_at)
        )

    @log_action
    def view_tasks(self):
        rows = self.db.execute("SELECT * FROM tasks").fetchall()
        return rows

    @log_action
    def delete_task(self, task_id: int):
        self.db.execute("DELETE FROM tasks WHERE id = ?", (task_id,))

    @log_action
    def update_status(self, task_id: int, status: str):
        self.db.execute(
            "UPDATE tasks SET status = ? WHERE id = ?",
            (status, task_id)
        )


# ---------------- THREADING (Background Logger) ----------------
def background_logger():
    while True:
        logging.info("System running in background...")
        threading.Event().wait(10)


# ---------------- CLI APPLICATION ----------------
class TaskApp:
    def __init__(self):
        self.db = Database()
        self.service = TaskService(self.db)

    def menu(self):
        while True:
            print("\n===== TASK MANAGER =====")
            print("1. Add Task")
            print("2. View Tasks")
            print("3. Delete Task")
            print("4. Update Status")
            print("5. Exit")

            choice = input("Enter choice: ")

            if choice == "1":
                title = input("Task title: ")
                self.service.add_task(Task(title))
                print("Task added.")

            elif choice == "2":
                tasks = self.service.view_tasks()
                for t in tasks:
                    print(t)

            elif choice == "3":
                task_id = int(input("Task ID: "))
                self.service.delete_task(task_id)
                print("Deleted.")

            elif choice == "4":
                task_id = int(input("Task ID: "))
                status = input("New status: ")
                self.service.update_status(task_id, status)
                print("Updated.")

            elif choice == "5":
                print("Exiting...")
                break

            else:
                print("Invalid choice")


# ---------------- MAIN ----------------
if __name__ == "__main__":
    # start background thread
    t = threading.Thread(target=background_logger, daemon=True)
    t.start()

    # run app
    app = TaskApp()
    app.menu()
```

```python
from functools import wraps
import time

def retry(max_attempts=3):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            attempts = 0

            while attempts < max_attempts:
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    attempts += 1
                    print(f"Attempt {attempts} failed: {e}")

            raise RuntimeError("Maximum retry attempts reached")

        return wrapper
    return decorator

@retry(max_attempts=5)
def unstable_function():
    import random
    if random.random() < 0.8:
        raise ValueError("Random failure")
    return "Success"

print(unstable_function())
```

```python
import asyncio
from dataclasses import dataclass
from typing import List
import aiohttp


@dataclass
class FetchResult:
    url: str
    status: int
    content_length: int


class AsyncWebFetcher:
    def __init__(self, max_concurrent: int = 5):
        self.semaphore = asyncio.Semaphore(max_concurrent)

    async def fetch(self, session: aiohttp.ClientSession, url: str) -> FetchResult:
        async with self.semaphore:
            try:
                async with session.get(url, timeout=10) as response:
                    content = await response.text()
                    return FetchResult(
                        url=url,
                        status=response.status,
                        content_length=len(content)
                    )
            except Exception:
                return FetchResult(
                    url=url,
                    status=-1,
                    content_length=0
                )

    async def run(self, urls: List[str]) -> List[FetchResult]:
        async with aiohttp.ClientSession() as session:
            tasks = [self.fetch(session, url) for url in urls]
            return await asyncio.gather(*tasks)


async def main():
    urls = [
        "https://python.org",
        "https://github.com",
        "https://stackoverflow.com",
        "https://example.com"
    ]

    fetcher = AsyncWebFetcher(max_concurrent=3)
    results = await fetcher.run(urls)

    for result in results:
        print(
            f"{result.url:<30} "
            f"Status: {result.status:<3} "
            f"Length: {result.content_length}"
        )


if __name__ == "__main__":
    asyncio.run(main())
```

```python
import json
from datetime import datetime


class BankAccount:
    def __init__(self, account_no, holder_name, balance=0):
        self.account_no = account_no
        self.holder_name = holder_name
        self.__balance = balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Amount must be positive.")

        self.__balance += amount
        print(f"৳{amount} deposited successfully.")

    def withdraw(self, amount):
        if amount > self.__balance:
            raise ValueError("Insufficient balance.")

        self.__balance -= amount
        print(f"৳{amount} withdrawn successfully.")

    def get_balance(self):
        return self.__balance

    def to_dict(self):
        return {
            "account_no": self.account_no,
            "holder_name": self.holder_name,
            "balance": self.__balance
        }


class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self):
        account_no = input("Enter Account Number: ")

        if account_no in self.accounts:
            print("Account already exists.")
            return

        name = input("Enter Holder Name: ")
        balance = float(input("Initial Deposit: "))

        self.accounts[account_no] = BankAccount(
            account_no,
            name,
            balance
        )

        print("Account created successfully.")

    def find_account(self, account_no):
        return self.accounts.get(account_no)

    def transfer_money(self):
        sender = input("Sender Account: ")
        receiver = input("Receiver Account: ")
        amount = float(input("Amount: "))

        sender_acc = self.find_account(sender)
        receiver_acc = self.find_account(receiver)

        if not sender_acc or not receiver_acc:
            print("Invalid account number.")
            return

        try:
            sender_acc.withdraw(amount)
            receiver_acc.deposit(amount)

            print("Transfer Successful")
            print(
                f"Time: {datetime.now().strftime('%d-%m-%Y %H:%M:%S')}"
            )

        except ValueError as e:
            print("Error:", e)

    def show_account(self):
        account_no = input("Account Number: ")

        account = self.find_account(account_no)

        if account:
            print("\n----- ACCOUNT INFO -----")
            print("Account No :", account.account_no)
            print("Name       :", account.holder_name)
            print("Balance    :", account.get_balance())
        else:
            print("Account not found.")

    def save_data(self):
        data = {
            acc_no: acc.to_dict()
            for acc_no, acc in self.accounts.items()
        }

        with open("accounts.json", "w") as file:
            json.dump(data, file, indent=4)

        print("Data saved successfully.")

    def load_data(self):
        try:
            with open("accounts.json", "r") as file:
                data = json.load(file)

            for acc_no, acc_data in data.items():
                self.accounts[acc_no] = BankAccount(
                    acc_data["account_no"],
                    acc_data["holder_name"],
                    acc_data["balance"]
                )

            print("Data loaded successfully.")

        except FileNotFoundError:
            print("No previous data found.")


def main():
    bank = Bank()
    bank.load_data()

    while True:
        print("\n====== BANK MANAGEMENT SYSTEM ======")
        print("1. Create Account")
        print("2. Transfer Money")
        print("3. Show Account")
        print("4. Save Data")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            bank.create_account()

        elif choice == "2":
            bank.transfer_money()

        elif choice == "3":
            bank.show_account()

        elif choice == "4":
            bank.save_data()

        elif choice == "5":
            bank.save_data()
            print("Thank you for using the system.")
            break

        else:
            print("Invalid choice.")


if __name__ == "__main__":
    main()
```

```python
from datetime import datetime


class InsufficientBalanceError(Exception):
    pass


class BankAccount:
    bank_name = "Future Bank"

    def __init__(self, account_number, owner, balance=0):
        self.account_number = account_number
        self.owner = owner
        self.__balance = balance
        self.transactions = []

    @property
    def balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError("Deposit amount must be positive.")

        self.__balance += amount
        self.transactions.append(
            f"{datetime.now()} | Deposited: ${amount}"
        )

    def withdraw(self, amount):
        if amount > self.__balance:
            raise InsufficientBalanceError(
                "Not enough balance."
            )

        self.__balance -= amount
        self.transactions.append(
            f"{datetime.now()} | Withdrawn: ${amount}"
        )

    def transfer(self, target_account, amount):
        self.withdraw(amount)
        target_account.deposit(amount)

        self.transactions.append(
            f"{datetime.now()} | Sent ${amount} to {target_account.owner}"
        )

        target_account.transactions.append(
            f"{datetime.now()} | Received ${amount} from {self.owner}"
        )

    def show_transactions(self):
        print("\nTransaction History")
        print("-" * 50)

        for transaction in self.transactions:
            print(transaction)

    def __str__(self):
        return (
            f"Account: {self.account_number}\n"
            f"Owner: {self.owner}\n"
            f"Balance: ${self.__balance}"
        )


class SavingsAccount(BankAccount):
    def __init__(
        self,
        account_number,
        owner,
        balance=0,
        interest_rate=5
    ):
        super().__init__(account_number, owner, balance)
        self.interest_rate = interest_rate

    def add_interest(self):
        interest = self.balance * self.interest_rate / 100
        self.deposit(interest)

        self.transactions.append(
            f"{datetime.now()} | Interest Added: ${interest:.2f}"
        )


class CurrentAccount(BankAccount):
    def __init__(
        self,
        account_number,
        owner,
        balance=0,
        overdraft_limit=500
    ):
        super().__init__(account_number, owner, balance)
        self.overdraft_limit = overdraft_limit

    def withdraw(self, amount):
        if amount > self.balance + self.overdraft_limit:
            raise InsufficientBalanceError(
                "Overdraft limit exceeded."
            )

        self._BankAccount__balance -= amount

        self.transactions.append(
            f"{datetime.now()} | Withdrawn: ${amount}"
        )


# --------------------
# Demo
# --------------------

account1 = SavingsAccount(
    "ACC101",
    "Kalpo",
    10000,
    10
)

account2 = CurrentAccount(
    "ACC102",
    "Sarah",
    5000,
    1000
)

try:
    account1.deposit(3000)
    account1.withdraw(1500)

    account1.transfer(account2, 2000)

    account1.add_interest()

except Exception as e:
    print("Error:", e)

print("\n")
print(account1)

print("\n")
print(account2)

print("\n")
account1.show_transactions()

print("\n")
account2.show_transactions()
```

```python
import numpy as np
import matplotlib.pyplot as plt


class LinearRegression:

    def __init__(self, learning_rate=0.01, epochs=1000):
        self.learning_rate = learning_rate
        self.epochs = epochs
        self.weight = 0
        self.bias = 0

    def fit(self, X, y):

        n = len(X)

        for epoch in range(self.epochs):

            y_pred = self.weight * X + self.bias

            dw = (-2 / n) * np.sum(X * (y - y_pred))
            db = (-2 / n) * np.sum(y - y_pred)

            self.weight -= self.learning_rate * dw
            self.bias -= self.learning_rate * db

            if epoch % 100 == 0:
                loss = np.mean((y - y_pred) ** 2)
                print(f"Epoch {epoch} | Loss: {loss:.4f}")

    def predict(self, X):
        return self.weight * X + self.bias


# Sample Dataset
X = np.array([1, 2, 3, 4, 5, 6, 7, 8])
y = np.array([3, 5, 7, 9, 11, 13, 15, 17])

model = LinearRegression(
    learning_rate=0.01,
    epochs=1000
)

model.fit(X, y)

predictions = model.predict(X)

print("\nWeight:", model.weight)
print("Bias:", model.bias)

plt.scatter(X, y)
plt.plot(X, predictions)
plt.title("Linear Regression From Scratch")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
import time
import heapq
import threading
from datetime import datetime


class Task:
    def __init__(self, name, priority, delay, action):
        self.name = name
        self.priority = priority
        self.delay = delay  # seconds
        self.action = action
        self.scheduled_time = time.time() + delay

    def __lt__(self, other):
        # higher priority runs first, if same priority earlier time runs first
        if self.priority == other.priority:
            return self.scheduled_time < other.scheduled_time
        return self.priority > other.priority


class TaskScheduler:
    def __init__(self):
        self.queue = []
        self.running = False

    def add_task(self, task):
        heapq.heappush(self.queue, task)
        print(f"[{datetime.now()}] Task added: {task.name} (P{task.priority})")

    def run_task(self, task):
        print(f"\nExecuting: {task.name}")
        print(f"Priority: {task.priority}")

        try:
            task.action()
        except Exception as e:
            print(f"Error in {task.name}: {e}")

        print(f"Finished: {task.name}\n")

    def start(self):
        self.running = True

        while self.running and self.queue:

            task = heapq.heappop(self.queue)

            # wait until scheduled time
            wait_time = task.scheduled_time - time.time()
            if wait_time > 0:
                time.sleep(wait_time)

            # run task in separate thread
            thread = threading.Thread(target=self.run_task, args=(task,))
            thread.start()
            thread.join()

    def stop(self):
        self.running = False


# ----------------------------
# Example Actions
# ----------------------------

def task_download():
    print("Downloading file...")
    time.sleep(1)
    print("Download complete!")

def task_email():
    print("Sending email...")
    time.sleep(1)
    print("Email sent!")

def task_backup():
    print("Backup started...")
    time.sleep(2)
    print("Backup completed!")


# ----------------------------
# Demo
# ----------------------------

scheduler = TaskScheduler()

scheduler.add_task(Task("Low Priority Backup", 1, 2, task_backup))
scheduler.add_task(Task("High Priority Email", 3, 1, task_email))
scheduler.add_task(Task("Medium Priority Download", 2, 3, task_download))

scheduler.start()
```

```python
import time
from collections import defaultdict


class Table:
    def __init__(self, name, schema):
        self.name = name
        self.schema = schema
        self.rows = []
        self.indexes = defaultdict(dict)  # column -> {value: [rows]}

    def _update_indexes(self, row):
        for col in self.schema:
            if col in row:
                self.indexes[col].setdefault(row[col], []).append(row)

    def insert(self, row):
        # validation
        for col in row:
            if col not in self.schema:
                raise ValueError(f"Invalid column: {col}")

        self.rows.append(row)
        self._update_indexes(row)

        print(f"[INSERT] {row}")

    def select(self, where=None):
        result = self.rows

        if where:
            key, value = where

            # use index if available
            if key in self.indexes:
                result = self.indexes[key].get(value, [])
            else:
                result = [r for r in self.rows if r.get(key) == value]

        return result

    def delete(self, where):
        key, value = where

        new_rows = []
        for row in self.rows:
            if row.get(key) != value:
                new_rows.append(row)

        self.rows = new_rows
        self.rebuild_indexes()

        print(f"[DELETE] where {key}={value}")

    def update(self, where, updates):
        key, value = where

        for row in self.rows:
            if row.get(key) == value:
                for k, v in updates.items():
                    row[k] = v

        self.rebuild_indexes()
        print(f"[UPDATE] where {key}={value} -> {updates}")

    def rebuild_indexes(self):
        self.indexes = defaultdict(dict)
        for row in self.rows:
            self._update_indexes(row)

    def show_all(self):
        print("\nTABLE DATA:")
        for row in self.rows:
            print(row)


class Database:
    def __init__(self):
        self.tables = {}

    def create_table(self, name, schema):
        self.tables[name] = Table(name, schema)
        print(f"Table '{name}' created with schema {schema}")

    def get_table(self, name):
        return self.tables.get(name)


# -------------------------
# Demo Usage
# -------------------------

db = Database()

db.create_table("users", ["id", "name", "age"])

users = db.get_table("users")

# Insert data
users.insert({"id": 1, "name": "Kalpo", "age": 25})
users.insert({"id": 2, "name": "Sarah", "age": 22})
users.insert({"id": 3, "name": "John", "age": 25})

# Select
print("\nSELECT age=25:")
print(users.select(where=("age", 25)))

# Update
users.update(where=("id", 2), updates={"age": 23})

# Delete
users.delete(where=("name", "John"))

# Show final data
users.show_all()
```

```python
import socket
import threading
import json


class Request:
    def __init__(self, method, path, body):
        self.method = method
        self.path = path
        self.body = body


class Response:
    def __init__(self, data, status=200):
        self.data = data
        self.status = status

    def to_http(self):
        response_body = json.dumps(self.data)

        return (
            f"HTTP/1.1 {self.status} OK\r\n"
            f"Content-Type: application/json\r\n"
            f"Content-Length: {len(response_body)}\r\n"
            f"\r\n"
            f"{response_body}"
        )


class MiniServer:
    def __init__(self, host="127.0.0.1", port=8080):
        self.host = host
        self.port = port
        self.routes = {}

    def route(self, path):
        def wrapper(func):
            self.routes[path] = func
            return func
        return wrapper

    def parse_request(self, data):
        lines = data.split("\r\n")
        request_line = lines[0].split()

        method = request_line[0]
        path = request_line[1]

        body = None
        if lines[-1]:
            try:
                body = json.loads(lines[-1])
            except:
                body = lines[-1]

        return Request(method, path, body)

    def handle_client(self, client_socket):
        data = client_socket.recv(1024).decode()

        if not data:
            client_socket.close()
            return

        request = self.parse_request(data)

        if request.path in self.routes:
            response = self.routes[request.path](request)
        else:
            response = Response({"error": "Not Found"}, 404)

        client_socket.send(response.to_http().encode())
        client_socket.close()

    def start(self):
        server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        server.bind((self.host, self.port))
        server.listen(5)

        print(f"Server running at http://{self.host}:{self.port}")

        while True:
            client_socket, addr = server.accept()

            thread = threading.Thread(
                target=self.handle_client,
                args=(client_socket,)
            )
            thread.start()


# -----------------------
# Create Server
# -----------------------

app = MiniServer()


@app.route("/")
def home(req):
    return Response({
        "message": "Welcome to Mini Python Server",
        "status": "running"
    })


@app.route("/add")
def add(req):
    data = req.body or {}

    a = data.get("a", 0)
    b = data.get("b", 0)

    return Response({
        "result": a + b
    })


@app.route("/user")
def user(req):
    return Response({
        "name": "Kalpo",
        "role": "Developer",
        "framework": "MiniPythonServer"
    })


# -----------------------
# Run Server
# -----------------------

if __name__ == "__main__":
    app.start()
```

```python
import hashlib
import time
import json


class Block:
    def __init__(self, index, transactions, previous_hash, difficulty=4):
        self.index = index
        self.timestamp = time.time()
        self.transactions = transactions
        self.previous_hash = previous_hash
        self.nonce = 0
        self.difficulty = difficulty
        self.hash = self.mine_block()

    def calculate_hash(self):
        data = {
            "index": self.index,
            "timestamp": self.timestamp,
            "transactions": self.transactions,
            "previous_hash": self.previous_hash,
            "nonce": self.nonce
        }

        block_string = json.dumps(data, sort_keys=True).encode()
        return hashlib.sha256(block_string).hexdigest()

    def mine_block(self):
        target = "0" * self.difficulty

        while True:
            hash_val = self.calculate_hash()

            if hash_val.startswith(target):
                return hash_val

            self.nonce += 1


class Blockchain:
    def __init__(self):
        self.chain = []
        self.pending_transactions = []

        # genesis block
        genesis_block = Block(0, ["Genesis Block"], "0")
        self.chain.append(genesis_block)

    def add_transaction(self, sender, receiver, amount):
        self.pending_transactions.append({
            "sender": sender,
            "receiver": receiver,
            "amount": amount
        })

    def mine_pending_transactions(self, miner_address):
        reward_tx = {
            "sender": "SYSTEM",
            "receiver": miner_address,
            "amount": 10
        }

        self.pending_transactions.append(reward_tx)

        new_block = Block(
            index=len(self.chain),
            transactions=self.pending_transactions,
            previous_hash=self.chain[-1].hash
        )

        self.chain.append(new_block)
        self.pending_transactions = []

        print(f"Block mined! Hash: {new_block.hash}")

    def is_chain_valid(self):
        for i in range(1, len(self.chain)):
            current = self.chain[i]
            previous = self.chain[i - 1]

            if current.hash != current.calculate_hash():
                return False

            if current.previous_hash != previous.hash:
                return False

        return True

    def print_chain(self):
        for block in self.chain:
            print("\n------------------------")
            print(f"Index: {block.index}")
            print(f"Hash: {block.hash}")
            print(f"Prev Hash: {block.previous_hash}")
            print(f"Transactions: {block.transactions}")


# ------------------------
# Demo
# ------------------------

blockchain = Blockchain()

blockchain.add_transaction("Kalpo", "Sarah", 50)
blockchain.add_transaction("John", "Kalpo", 30)

print("Mining block...")
blockchain.mine_pending_transactions("Kalpo")

blockchain.add_transaction("Kalpo", "Alice", 20)
blockchain.add_transaction("Bob", "Kalpo", 10)

print("Mining second block...")
blockchain.mine_pending_transactions("Kalpo")

print("\nChain valid?", blockchain.is_chain_valid())

blockchain.print_chain()
```

```python
import time
from collections import deque


class Process:
    def __init__(self, pid, burst_time, priority=1):
        self.pid = pid
        self.burst_time = burst_time
        self.remaining_time = burst_time
        self.priority = priority
        self.waiting_time = 0
        self.turnaround_time = 0


class CPUScheduler:
    def __init__(self, time_quantum=2):
        self.queue = deque()
        self.time_quantum = time_quantum
        self.time = 0

    def add_process(self, process):
        self.queue.append(process)
        print(f"Added Process {process.pid} | Burst: {process.burst_time}")

    def run_round_robin(self):
        print("\n--- Round Robin Scheduling Started ---\n")

        while self.queue:
            process = self.queue.popleft()

            exec_time = min(self.time_quantum, process.remaining_time)

            print(f"Executing P{process.pid} for {exec_time}s")
            time.sleep(0.2)  # simulation delay

            process.remaining_time -= exec_time
            self.time += exec_time

            if process.remaining_time > 0:
                self.queue.append(process)
            else:
                process.turnaround_time = self.time
                process.waiting_time = process.turnaround_time - process.burst_time

                print(f"P{process.pid} finished | WT={process.waiting_time} TT={process.turnaround_time}")

    def priority_scheduling(self, processes):
        print("\n--- Priority Scheduling Started ---\n")

        processes.sort(key=lambda x: x.priority, reverse=True)

        current_time = 0

        for p in processes:
            print(f"Running P{p.pid} with priority {p.priority}")
            time.sleep(0.2)

            current_time += p.burst_time
            p.turnaround_time = current_time
            p.waiting_time = current_time - p.burst_time

            print(f"P{p.pid} finished | WT={p.waiting_time} TT={p.turnaround_time}")

    def show_summary(self, processes):
        print("\n--- Summary ---")
        for p in processes:
            print(
                f"P{p.pid} | BT={p.burst_time} | WT={p.waiting_time} | TT={p.turnaround_time}"
            )


# -------------------------
# Demo
# -------------------------

scheduler = CPUScheduler(time_quantum=3)

p1 = Process(1, 10)
p2 = Process(2, 5)
p3 = Process(3, 8)

scheduler.add_process(p1)
scheduler.add_process(p2)
scheduler.add_process(p3)

scheduler.run_round_robin()

# Priority scheduling demo
p4 = Process(4, 6, priority=2)
p5 = Process(5, 4, priority=5)
p6 = Process(6, 7, priority=3)

scheduler.priority_scheduling([p4, p5, p6])
scheduler.show_summary([p4, p5, p6])
```

```python
import math
from collections import defaultdict, Counter


class MiniSearchEngine:
    def __init__(self):
        self.documents = {}
        self.inverted_index = defaultdict(set)
        self.doc_freq = defaultdict(int)

    def add_document(self, doc_id, text):
        words = text.lower().split()
        self.documents[doc_id] = words

        unique_words = set(words)

        for word in words:
            self.inverted_index[word].add(doc_id)

        for word in unique_words:
            self.doc_freq[word] += 1

    def tf(self, word, doc):
        return doc.count(word) / len(doc)

    def idf(self, word):
        total_docs = len(self.documents)
        return math.log((total_docs + 1) / (self.doc_freq[word] + 1)) + 1

    def score(self, query, doc_id):
        doc = self.documents[doc_id]
        score = 0

        for word in query:
            tf = self.tf(word, doc)
            idf = self.idf(word)
            score += tf * idf

        return score

    def search(self, query):
        query_words = query.lower().split()

        candidate_docs = set()

        for word in query_words:
            candidate_docs.update(self.inverted_index.get(word, []))

        ranked = []

        for doc_id in candidate_docs:
            s = self.score(query_words, doc_id)
            ranked.append((s, doc_id))

        ranked.sort(reverse=True)

        return ranked


# ------------------------
# Demo
# ------------------------

engine = MiniSearchEngine()

engine.add_document(1, "python is a powerful programming language")
engine.add_document(2, "python supports machine learning and ai")
engine.add_document(3, "java is also a programming language")
engine.add_document(4, "machine learning is part of ai")
engine.add_document(5, "python is widely used in data science")

results = engine.search("python machine learning")

print("\nSearch Results:")
for score, doc_id in results:
    print(f"Doc {doc_id} | Score: {score:.4f} | Text: {' '.join(engine.documents[doc_id])}")
```

```python
class VM:
    def __init__(self):
        self.stack = []
        self.variables = {}

    def run(self, instructions):
        ip = 0  # instruction pointer

        while ip < len(instructions):
            instr = instructions[ip]
            op = instr[0]

            # PUSH value
            if op == "PUSH":
                self.stack.append(instr[1])

            # ADD
            elif op == "ADD":
                b = self.stack.pop()
                a = self.stack.pop()
                self.stack.append(a + b)

            # SUB
            elif op == "SUB":
                b = self.stack.pop()
                a = self.stack.pop()
                self.stack.append(a - b)

            # MUL
            elif op == "MUL":
                b = self.stack.pop()
                a = self.stack.pop()
                self.stack.append(a * b)

            # DIV
            elif op == "DIV":
                b = self.stack.pop()
                a = self.stack.pop()
                self.stack.append(a / b)

            # STORE variable
            elif op == "STORE":
                value = self.stack.pop()
                self.variables[instr[1]] = value

            # LOAD variable
            elif op == "LOAD":
                self.stack.append(self.variables[instr[1]])

            # PRINT
            elif op == "PRINT":
                print("OUTPUT:", self.stack[-1])

            # JMP (jump)
            elif op == "JMP":
                ip = instr[1]
                continue

            # JMP_IF_ZERO
            elif op == "JMP_IF_ZERO":
                value = self.stack.pop()
                if value == 0:
                    ip = instr[1]
                    continue

            ip += 1


# -----------------------
# Demo Program
# -----------------------

vm = VM()

program = [
    ("PUSH", 10),
    ("PUSH", 20),
    ("ADD",),
    ("STORE", "result"),

    ("LOAD", "result"),
    ("PUSH", 5),
    ("MUL",),

    ("PRINT",),

    # loop style simulation
    ("PUSH", 3),
    ("STORE", "counter"),

    ("LOAD", "counter"),
    ("PUSH", 1),
    ("SUB",),
    ("STORE", "counter"),

    ("LOAD", "counter"),
    ("PRINT",),
]

vm.run(program)
```

```python
import time


class Object:
    def __init__(self, name):
        self.name = name
        self.references = 0

    def __repr__(self):
        return f"{self.name} (refs={self.references})"


class MemoryManager:
    def __init__(self):
        self.objects = {}

    def create(self, name):
        obj = Object(name)
        self.objects[name] = obj
        print(f"[CREATE] {obj}")
        return obj

    def add_reference(self, name):
        if name in self.objects:
            self.objects[name].references += 1
            print(f"[REF +] {self.objects[name]}")

    def remove_reference(self, name):
        if name in self.objects:
            obj = self.objects[name]
            obj.references -= 1
            print(f"[REF -] {obj}")

            if obj.references <= 0:
                self.collect(name)

    def collect(self, name):
        if name in self.objects:
            print(f"[GC] Collecting {name}")
            del self.objects[name]

    def status(self):
        print("\n--- MEMORY STATUS ---")
        for obj in self.objects.values():
            print(obj)
        print("---------------------\n")


# -----------------------
# Demo
# -----------------------

mm = MemoryManager()

a = mm.create("ObjectA")
b = mm.create("ObjectB")
c = mm.create("ObjectC")

mm.add_reference("ObjectA")
mm.add_reference("ObjectA")
mm.add_reference("ObjectB")

mm.status()

mm.remove_reference("ObjectA")
mm.remove_reference("ObjectA")
mm.remove_reference("ObjectB")

mm.status()

mm.remove_reference("ObjectC")  # no refs → GC trigger
mm.status()
```

```python
class AsyncPipeline:
    def __init__(self):
        self.tasks = []

    def task(self, func):
        self.tasks.append(func)
        return func

    async def run(self, data):
        result = data
        for task in self.tasks:
            result = await task(result)
        return result


import asyncio

pipeline = AsyncPipeline()

@pipeline.task
async def double(x):
    await asyncio.sleep(0.1)
    return x * 2

@pipeline.task
async def square(x):
    await asyncio.sleep(0.1)
    return x ** 2

async def main():
    result = await pipeline.run(5)
    print(f"Final Result: {result}")

asyncio.run(main())
```

```python
from collections import OrderedDict
from functools import wraps

class LRUCache:
    def __init__(self, capacity=5):
        self.capacity = capacity
        self.cache = OrderedDict()

    def __call__(self, func):
        @wraps(func)
        def wrapper(*args):
            if args in self.cache:
                self.cache.move_to_end(args)
                return self.cache[args]

            result = func(*args)
            self.cache[args] = result

            if len(self.cache) > self.capacity:
                self.cache.popitem(last=False)

            return result
        return wrapper


@LRUCache(capacity=3)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(35))
```

```python
from functools import wraps

def memoize(func):
    cache = {}

    @wraps(func)
    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]

    return wrapper


@memoize
def factorial(n):
    if n == 0:
        return 1
    return n * factorial(n - 1)

print(factorial(500))
```

```python
import re

def password_strength(password):
    score = 0

    if len(password) >= 8:
        score += 1
    if re.search(r"[A-Z]", password):
        score += 1
    if re.search(r"[a-z]", password):
        score += 1
    if re.search(r"\d", password):
        score += 1
    if re.search(r"[!@#$%^&*()_+]", password):
        score += 1

    levels = {
        1: "Very Weak",
        2: "Weak",
        3: "Medium",
        4: "Strong",
        5: "Very Strong"
    }

    return levels.get(score, "Very Weak")

print(password_strength("MyPass123!"))
```

```python
import asyncio
import time
from collections import OrderedDict


# =========================
# 🔥 LRU CACHE SYSTEM
# =========================
class LRUCache:
    def __init__(self, capacity=3):
        self.cache = OrderedDict()
        self.capacity = capacity

    def get(self, key):
        if key not in self.cache:
            return None
        self.cache.move_to_end(key)
        return self.cache[key]

    def put(self, key, value):
        if key in self.cache:
            self.cache.move_to_end(key)

        self.cache[key] = value

        if len(self.cache) > self.capacity:
            removed = self.cache.popitem(last=False)
            print(f"[CACHE] Evicted: {removed}")


# =========================
# 📡 EVENT SYSTEM
# =========================
class EventBus:
    def __init__(self):
        self.listeners = {}

    def on(self, event, callback):
        if event not in self.listeners:
            self.listeners[event] = []
        self.listeners[event].append(callback)

    def emit(self, event, data=None):
        if event in self.listeners:
            for callback in self.listeners[event]:
                callback(data)


# =========================
# 🌐 FAKE WEB FETCHER
# =========================
class FakeWeb:
    async def fetch(self, url):
        delay = len(url) % 3 + 1
        await asyncio.sleep(delay)

        return {
            "url": url,
            "content": f"Content_of_{url}",
            "timestamp": time.time()
        }


# =========================
# 🕷️ CRAWLER ENGINE
# =========================
class AsyncCrawler:
    def __init__(self, urls):
        self.urls = urls
        self.cache = LRUCache(capacity=2)
        self.bus = EventBus()
        self.web = FakeWeb()

        # events
        self.bus.on("fetch_start", self.on_start)
        self.bus.on("fetch_done", self.on_done)
        self.bus.on("cache_hit", self.on_cache_hit)

    # event handlers
    def on_start(self, url):
        print(f"[START] Fetching: {url}")

    def on_done(self, data):
        print(f"[DONE] {data['url']} -> {data['content']}")

    def on_cache_hit(self, url):
        print(f"[CACHE HIT] {url}")

    async def fetch_url(self, url):
        # cache check
        cached = self.cache.get(url)
        if cached:
            self.bus.emit("cache_hit", url)
            return cached

        self.bus.emit("fetch_start", url)

        data = await self.web.fetch(url)

        self.cache.put(url, data)

        self.bus.emit("fetch_done", data)
        return data

    async def run(self):
        tasks = [self.fetch_url(url) for url in self.urls]
        results = await asyncio.gather(*tasks)
        return results


# =========================
# 🚀 RUNNER
# =========================
if __name__ == "__main__":
    urls = [
        "google.com",
        "github.com",
        "openai.com",
        "stackoverflow.com",
        "github.com",   # duplicate to test cache
        "openai.com"    # duplicate to test cache
    ]

    crawler = AsyncCrawler(urls)

    asyncio.run(crawler.run())
```

```python
class BlockNegative(type):
    def __new__(cls, name, bases, dct):
        for key, value in dct.items():
            if callable(value):
                def wrapper(*args, **kwargs):
                    result = value(*args, **kwargs)
                    if isinstance(result, (int, float)) and result < 0:
                        raise ValueError("Negative result not allowed")
                    return result
                dct[key] = wrapper

        return super().__new__(cls, name, bases, dct)


class Calculator(metaclass=BlockNegative):
    def subtract(self, a, b):
        return a - b


c = Calculator()
print(c.subtract(5, 10))
```

```python
import asyncio

queue = asyncio.Queue()


async def producer():
    for i in range(5):
        await queue.put(i)
        print("Produced", i)
        await asyncio.sleep(1)


async def consumer():
    while True:
        item = await queue.get()
        print("Consumed", item)
        queue.task_done()


async def main():
    await asyncio.gather(producer(), consumer())


asyncio.run(main())
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            result = None
            for _ in range(n):
                result = func(*args, **kwargs)
            return result
        return wrapper
    return decorator


@repeat(3)
def greet(name):
    print("Hello", name)


greet("Kalpo")
```

```python
import random
from collections import defaultdict

class MarkovChain:
    def __init__(self):
        self.model = defaultdict(list)

    def train(self, text):
        words = text.split()

        for i in range(len(words) - 1):
            current_word = words[i]
            next_word = words[i + 1]
            self.model[current_word].append(next_word)

    def generate(self, start_word, length=20):
        word = start_word
        output = [word]

        for _ in range(length - 1):
            if word not in self.model:
                break

            word = random.choice(self.model[word])
            output.append(word)

        return " ".join(output)


# =========================
# SAMPLE TRAINING DATA
# =========================
text_data = """
football is a game of probability and skill
teams win based on performance and luck
world cup is the biggest football tournament
prediction models use probability and statistics
python can simulate random outcomes very well
"""

# =========================
# RUN MODEL
# =========================
mc = MarkovChain()
mc.train(text_data)

print("🔮 Generated Text:\n")
print(mc.generate("football", 25))
```

```python
import time

class Transaction:
    def __init__(self, t_type, amount, balance):
        self.t_type = t_type
        self.amount = amount
        self.balance = balance
        self.timestamp = time.strftime("%Y-%m-%d %H:%M:%S")

    def __str__(self):
        return f"[{self.timestamp}] {self.t_type}: {self.amount} | Balance: {self.balance}"


class Account:
    def __init__(self, name, initial_deposit=0):
        self.name = name
        self.balance = initial_deposit
        self.transactions = []
        self.add_transaction("Account Created", initial_deposit)

    def add_transaction(self, t_type, amount):
        t = Transaction(t_type, amount, self.balance)
        self.transactions.append(t)

    def deposit(self, amount):
        if amount <= 0:
            print("Invalid deposit amount")
            return
        self.balance += amount
        self.add_transaction("Deposit", amount)
        print(f"Deposited {amount}. New balance: {self.balance}")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid withdrawal amount")
            return
        if amount > self.balance:
            print("Insufficient balance!")
            return
        self.balance -= amount
        self.add_transaction("Withdraw", amount)
        print(f"Withdrawn {amount}. New balance: {self.balance}")

    def transfer(self, other_account, amount):
        if amount <= 0:
            print("Invalid transfer amount")
            return
        if amount > self.balance:
            print("Not enough balance to transfer")
            return

        self.balance -= amount
        other_account.balance += amount

        self.add_transaction(f"Transfer to {other_account.name}", amount)
        other_account.add_transaction(f"Transfer from {self.name}", amount)

        print(f"Transferred {amount} to {other_account.name}")

    def show_history(self):
        print(f"\n--- Transaction History of {self.name} ---")
        for t in self.transactions:
            print(t)

    def show_balance(self):
        print(f"{self.name}'s Balance: {self.balance}")


class Bank:
    def __init__(self):
        self.accounts = {}

    def create_account(self, name, deposit=0):
        if name in self.accounts:
            print("Account already exists!")
            return
        self.accounts[name] = Account(name, deposit)
        print(f"Account created for {name}")

    def get_account(self, name):
        return self.accounts.get(name)

    def show_all_accounts(self):
        print("\n--- All Accounts ---")
        for name, acc in self.accounts.items():
            print(f"{name} -> Balance: {acc.balance}")


# ---------------- CLI SYSTEM ---------------- #

def main():
    bank = Bank()

    while True:
        print("\n===== MINI BANK SYSTEM =====")
        print("1. Create Account")
        print("2. Deposit")
        print("3. Withdraw")
        print("4. Transfer")
        print("5. Show Balance")
        print("6. Transaction History")
        print("7. Show All Accounts")
        print("8. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter name: ")
            deposit = float(input("Initial deposit: "))
            bank.create_account(name, deposit)

        elif choice == "2":
            name = input("Account name: ")
            acc = bank.get_account(name)
            if acc:
                amount = float(input("Deposit amount: "))
                acc.deposit(amount)

        elif choice == "3":
            name = input("Account name: ")
            acc = bank.get_account(name)
            if acc:
                amount = float(input("Withdraw amount: "))
                acc.withdraw(amount)

        elif choice == "4":
            from_acc = input("From account: ")
            to_acc = input("To account: ")

            acc1 = bank.get_account(from_acc)
            acc2 = bank.get_account(to_acc)

            if acc1 and acc2:
                amount = float(input("Transfer amount: "))
                acc1.transfer(acc2, amount)

        elif choice == "5":
            name = input("Account name: ")
            acc = bank.get_account(name)
            if acc:
                acc.show_balance()

        elif choice == "6":
            name = input("Account name: ")
            acc = bank.get_account(name)
            if acc:
                acc.show_history()

        elif choice == "7":
            bank.show_all_accounts()

        elif choice == "8":
            print("Exiting system...")
            break

        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```

```python
import json
import os

class Student:
    def __init__(self, roll, name):
        self.roll = roll
        self.name = name
        self.marks = {}
        self.total = 0
        self.grade = ""

    def add_marks(self, subject, mark):
        self.marks[subject] = mark
        self.calculate_result()

    def calculate_result(self):
        self.total = sum(self.marks.values())
        avg = self.total / len(self.marks) if self.marks else 0

        if avg >= 80:
            self.grade = "A+"
        elif avg >= 70:
            self.grade = "A"
        elif avg >= 60:
            self.grade = "B"
        elif avg >= 50:
            self.grade = "C"
        else:
            self.grade = "F"

    def to_dict(self):
        return {
            "roll": self.roll,
            "name": self.name,
            "marks": self.marks,
            "total": self.total,
            "grade": self.grade
        }


class StudentSystem:
    def __init__(self, file_name="students.json"):
        self.file_name = file_name
        self.students = {}
        self.load_data()

    def add_student(self, roll, name):
        if roll in self.students:
            print("Student already exists!")
            return
        self.students[roll] = Student(roll, name)
        print("Student added successfully!")

    def add_marks(self, roll, subject, mark):
        student = self.students.get(roll)
        if not student:
            print("Student not found!")
            return
        student.add_marks(subject, mark)
        print("Marks added successfully!")

    def show_student(self, roll):
        student = self.students.get(roll)
        if not student:
            print("Not found!")
            return

        print("\n--- Student Info ---")
        print("Roll:", student.roll)
        print("Name:", student.name)
        print("Marks:", student.marks)
        print("Total:", student.total)
        print("Grade:", student.grade)

    def show_all(self):
        print("\n--- All Students ---")

        sorted_students = sorted(
            self.students.values(),
            key=lambda s: s.total,
            reverse=True
        )

        for i, s in enumerate(sorted_students, start=1):
            print(f"{i}. {s.name} | Roll: {s.roll} | Total: {s.total} | Grade: {s.grade}")

    def search_student(self, keyword):
        print("\n--- Search Result ---")
        found = False

        for s in self.students.values():
            if keyword.lower() in s.name.lower() or keyword == s.roll:
                print(f"{s.name} | Roll: {s.roll} | Total: {s.total}")
                found = True

        if not found:
            print("No student found!")

    def save_data(self):
        data = {roll: s.to_dict() for roll, s in self.students.items()}
        with open(self.file_name, "w") as f:
            json.dump(data, f, indent=4)
        print("Data saved!")

    def load_data(self):
        if not os.path.exists(self.file_name):
            return

        with open(self.file_name, "r") as f:
            data = json.load(f)

        for roll, info in data.items():
            s = Student(info["roll"], info["name"])
            s.marks = info["marks"]
            s.total = info["total"]
            s.grade = info["grade"]
            self.students[roll] = s


# ---------------- MENU SYSTEM ---------------- #

def main():
    system = StudentSystem()

    while True:
        print("\n===== STUDENT MANAGEMENT SYSTEM =====")
        print("1. Add Student")
        print("2. Add Marks")
        print("3. Show Student")
        print("4. Show All Students (Ranking)")
        print("5. Search Student")
        print("6. Save Data")
        print("7. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            roll = input("Roll: ")
            name = input("Name: ")
            system.add_student(roll, name)

        elif choice == "2":
            roll = input("Roll: ")
            subject = input("Subject: ")
            mark = float(input("Marks: "))
            system.add_marks(roll, subject, mark)

        elif choice == "3":
            roll = input("Roll: ")
            system.show_student(roll)

        elif choice == "4":
            system.show_all()

        elif choice == "5":
            keyword = input("Enter name or roll: ")
            system.search_student(keyword)

        elif choice == "6":
            system.save_data()

        elif choice == "7":
            system.save_data()
            print("Exiting...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

```python
import json
import os
from datetime import datetime, timedelta

class Book:
    def __init__(self, book_id, title, author, copies):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.copies = copies
        self.issued_to = {}  # roll -> due_date

    def to_dict(self):
        return {
            "book_id": self.book_id,
            "title": self.title,
            "author": self.author,
            "copies": self.copies,
            "issued_to": self.issued_to
        }


class Library:
    def __init__(self, file_name="library.json"):
        self.file_name = file_name
        self.books = {}
        self.load_data()

    # ---------------- BOOK MANAGEMENT ---------------- #

    def add_book(self, book_id, title, author, copies):
        if book_id in self.books:
            print("Book already exists! Updating copies...")
            self.books[book_id].copies += copies
            return

        self.books[book_id] = Book(book_id, title, author, copies)
        print("Book added successfully!")

    def show_books(self):
        print("\n--- Library Books ---")
        for b in self.books.values():
            issued = len(b.issued_to)
            available = b.copies - issued
            print(f"{b.book_id} | {b.title} | {b.author} | Available: {available}")

    def search_book(self, keyword):
        print("\n--- Search Result ---")
        found = False

        for b in self.books.values():
            if keyword.lower() in b.title.lower() or keyword.lower() in b.author.lower():
                print(f"{b.book_id} | {b.title} | {b.author}")
                found = True

        if not found:
            print("No book found!")

    # ---------------- ISSUE SYSTEM ---------------- #

    def issue_book(self, book_id, student_id):
        book = self.books.get(book_id)
        if not book:
            print("Book not found!")
            return

        if len(book.issued_to) >= book.copies:
            print("No copies available!")
            return

        if student_id in book.issued_to:
            print("Student already issued this book!")
            return

        due_date = datetime.now() + timedelta(days=7)
        book.issued_to[student_id] = due_date.strftime("%Y-%m-%d")

        print(f"Book issued to {student_id}. Due date: {due_date.date()}")

    # ---------------- RETURN SYSTEM ---------------- #

    def return_book(self, book_id, student_id):
        book = self.books.get(book_id)
        if not book:
            print("Book not found!")
            return

        if student_id not in book.issued_to:
            print("This student did not issue this book!")
            return

        due_date = datetime.strptime(book.issued_to[student_id], "%Y-%m-%d")
        today = datetime.now()

        fine = 0
        if today > due_date:
            days_late = (today - due_date).days
            fine = days_late * 10  # 10 per day fine

        del book.issued_to[student_id]

        print(f"Book returned successfully. Fine: {fine} Taka")

    # ---------------- DATA HANDLING ---------------- #

    def save_data(self):
        data = {bid: b.to_dict() for bid, b in self.books.items()}
        with open(self.file_name, "w") as f:
            json.dump(data, f, indent=4)
        print("Data saved!")

    def load_data(self):
        if not os.path.exists(self.file_name):
            return

        with open(self.file_name, "r") as f:
            data = json.load(f)

        for bid, info in data.items():
            book = Book(
                info["book_id"],
                info["title"],
                info["author"],
                info["copies"]
            )
            book.issued_to = info["issued_to"]
            self.books[bid] = book


# ---------------- CLI MENU ---------------- #

def main():
    lib = Library()

    while True:
        print("\n===== LIBRARY SYSTEM =====")
        print("1. Add Book")
        print("2. Show Books")
        print("3. Search Book")
        print("4. Issue Book")
        print("5. Return Book")
        print("6. Save Data")
        print("7. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            book_id = input("Book ID: ")
            title = input("Title: ")
            author = input("Author: ")
            copies = int(input("Copies: "))
            lib.add_book(book_id, title, author, copies)

        elif choice == "2":
            lib.show_books()

        elif choice == "3":
            keyword = input("Enter title or author: ")
            lib.search_book(keyword)

        elif choice == "4":
            book_id = input("Book ID: ")
            student_id = input("Student ID: ")
            lib.issue_book(book_id, student_id)

        elif choice == "5":
            book_id = input("Book ID: ")
            student_id = input("Student ID: ")
            lib.return_book(book_id, student_id)

        elif choice == "6":
            lib.save_data()

        elif choice == "7":
            lib.save_data()
            print("Exiting system...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

```python
import json
import os
import time
from datetime import datetime

class Question:
    def __init__(self, q, options, answer):
        self.q = q
        self.options = options
        self.answer = answer

    def is_correct(self, user_answer):
        return user_answer.lower() == self.answer.lower()


class Quiz:
    def __init__(self):
        self.questions = []
        self.score = 0

    def add_question(self, question):
        self.questions.append(question)

    def take_quiz(self, username):
        print(f"\nWelcome {username}! Starting Quiz...\n")
        self.score = 0

        start_time = time.time()

        for i, q in enumerate(self.questions, start=1):
            print(f"Q{i}: {q.q}")
            for idx, opt in enumerate(q.options, start=1):
                print(f"   {idx}. {opt}")

            try:
                choice = input("Your answer (number): ")

                if not choice.isdigit():
                    print("Invalid input!")
                    continue

                choice = int(choice)

                if choice < 1 or choice > len(q.options):
                    print("Out of range!")
                    continue

                selected = q.options[choice - 1]

                if q.is_correct(selected):
                    print("✅ Correct!\n")
                    self.score += 1
                else:
                    print(f"❌ Wrong! Correct answer: {q.answer}\n")

            except:
                print("Error occurred!")

        end_time = time.time()
        total_time = round(end_time - start_time, 2)

        print("\n===== QUIZ FINISHED =====")
        print(f"Score: {self.score}/{len(self.questions)}")
        print(f"Time Taken: {total_time} seconds")

        return self.score, total_time


class LeaderBoard:
    def __init__(self, file_name="leaderboard.json"):
        self.file_name = file_name
        self.data = self.load()

    def load(self):
        if not os.path.exists(self.file_name):
            return []
        with open(self.file_name, "r") as f:
            return json.load(f)

    def save(self):
        with open(self.file_name, "w") as f:
            json.dump(self.data, f, indent=4)

    def add_score(self, username, score, total_time):
        self.data.append({
            "name": username,
            "score": score,
            "time": total_time,
            "date": str(datetime.now())
        })
        self.save()

    def show(self):
        print("\n===== LEADERBOARD =====")
        sorted_data = sorted(self.data, key=lambda x: x["score"], reverse=True)

        for i, d in enumerate(sorted_data[:10], start=1):
            print(f"{i}. {d['name']} | Score: {d['score']} | Time: {d['time']}s")


# ---------------- QUIZ SYSTEM ---------------- #

def build_quiz():
    quiz = Quiz()

    quiz.add_question(Question(
        "What is the capital of Bangladesh?",
        ["Dhaka", "Chittagong", "Rajshahi", "Khulna"],
        "Dhaka"
    ))

    quiz.add_question(Question(
        "Which language is used for AI?",
        ["Python", "HTML", "CSS", "SQL"],
        "Python"
    ))

    quiz.add_question(Question(
        "2 + 2 * 2 = ?",
        ["6", "8", "4", "10"],
        "6"
    ))

    quiz.add_question(Question(
        "Who developed Python?",
        ["Guido van Rossum", "Elon Musk", "Bill Gates", "Mark Zuckerberg"],
        "Guido van Rossum"
    ))

    return quiz


def main():
    quiz = build_quiz()
    leaderboard = LeaderBoard()

    while True:
        print("\n===== QUIZ MENU =====")
        print("1. Start Quiz")
        print("2. View Leaderboard")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter your name: ")
            score, t = quiz.take_quiz(name)
            leaderboard.add_score(name, score, t)

        elif choice == "2":
            leaderboard.show()

        elif choice == "3":
            print("Exiting quiz system...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

```python
import json
import os
import time

class User:
    def __init__(self, name, pin, balance=0):
        self.name = name
        self.pin = pin
        self.balance = balance
        self.transactions = []

    def add_transaction(self, t_type, amount):
        self.transactions.append({
            "type": t_type,
            "amount": amount,
            "time": time.strftime("%Y-%m-%d %H:%M:%S"),
            "balance": self.balance
        })


class ATMSystem:
    def __init__(self, file_name="atm_data.json"):
        self.file_name = file_name
        self.users = {}
        self.load_data()

    # ---------------- USER MANAGEMENT ---------------- #

    def create_account(self, name, pin, deposit=0):
        if name in self.users:
            print("Account already exists!")
            return

        self.users[name] = User(name, pin, deposit)
        self.users[name].add_transaction("Account Created", deposit)
        print("Account created successfully!")

    def login(self, name, pin):
        user = self.users.get(name)
        if not user:
            print("User not found!")
            return None

        if user.pin != pin:
            print("Incorrect PIN!")
            return None

        print(f"Welcome {name}!")
        return user

    # ---------------- ATM OPERATIONS ---------------- #

    def deposit(self, user, amount):
        if amount <= 0:
            print("Invalid amount!")
            return

        user.balance += amount
        user.add_transaction("Deposit", amount)
        print(f"Deposited {amount}. New balance: {user.balance}")

    def withdraw(self, user, amount):
        if amount <= 0:
            print("Invalid amount!")
            return

        if amount > user.balance:
            print("Insufficient balance!")
            return

        # ATM withdrawal limit
        if amount > 20000:
            print("Daily limit exceeded (Max 20,000)")
            return

        user.balance -= amount
        user.add_transaction("Withdraw", amount)
        print(f"Withdrawn {amount}. Remaining balance: {user.balance}")

    def transfer(self, user, to_name, amount):
        if to_name not in self.users:
            print("Receiver not found!")
            return

        if amount <= 0:
            print("Invalid amount!")
            return

        if user.balance < amount:
            print("Not enough balance!")
            return

        receiver = self.users[to_name]

        user.balance -= amount
        receiver.balance += amount

        user.add_transaction(f"Transfer to {to_name}", amount)
        receiver.add_transaction(f"Transfer from {user.name}", amount)

        print(f"Transferred {amount} to {to_name}")

    def show_balance(self, user):
        print(f"Current Balance: {user.balance}")

    def mini_statement(self, user):
        print("\n===== MINI STATEMENT =====")
        for t in user.transactions[-5:]:
            print(f"{t['time']} | {t['type']} | {t['amount']} | Balance: {t['balance']}")

    # ---------------- DATA HANDLING ---------------- #

    def save_data(self):
        data = {}
        for name, user in self.users.items():
            data[name] = {
                "pin": user.pin,
                "balance": user.balance,
                "transactions": user.transactions
            }

        with open(self.file_name, "w") as f:
            json.dump(data, f, indent=4)

        print("Data saved successfully!")

    def load_data(self):
        if not os.path.exists(self.file_name):
            return

        with open(self.file_name, "r") as f:
            data = json.load(f)

        for name, info in data.items():
            user = User(name, info["pin"], info["balance"])
            user.transactions = info["transactions"]
            self.users[name] = user


# ---------------- MAIN MENU ---------------- #

def main():
    atm = ATMSystem()

    while True:
        print("\n===== ATM SYSTEM =====")
        print("1. Create Account")
        print("2. Login")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            name = input("Enter name: ")
            pin = input("Set 4-digit PIN: ")
            deposit = float(input("Initial deposit: "))
            atm.create_account(name, pin, deposit)

        elif choice == "2":
            name = input("Enter name: ")
            pin = input("Enter PIN: ")

            user = atm.login(name, pin)

            if user:
                while True:
                    print("\n--- ATM MENU ---")
                    print("1. Deposit")
                    print("2. Withdraw")
                    print("3. Transfer")
                    print("4. Balance")
                    print("5. Mini Statement")
                    print("6. Logout")

                    opt = input("Choose: ")

                    if opt == "1":
                        amt = float(input("Amount: "))
                        atm.deposit(user, amt)

                    elif opt == "2":
                        amt = float(input("Amount: "))
                        atm.withdraw(user, amt)

                    elif opt == "3":
                        to = input("Send to: ")
                        amt = float(input("Amount: "))
                        atm.transfer(user, to, amt)

                    elif opt == "4":
                        atm.show_balance(user)

                    elif opt == "5":
                        atm.mini_statement(user)

                    elif opt == "6":
                        print("Logging out...")
                        atm.save_data()
                        break

                    else:
                        print("Invalid option!")

        elif choice == "3":
            atm.save_data()
            print("Exiting ATM system...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

```python
import json
import os
import time

class Product:
    def __init__(self, pid, name, price, stock):
        self.pid = pid
        self.name = name
        self.price = price
        self.stock = stock

    def to_dict(self):
        return {
            "pid": self.pid,
            "name": self.name,
            "price": self.price,
            "stock": self.stock
        }


class Shop:
    def __init__(self, file_name="shop_data.json"):
        self.file_name = file_name
        self.products = {}
        self.cart = {}
        self.load_data()

    # ---------------- ADMIN ---------------- #

    def add_product(self, pid, name, price, stock):
        if pid in self.products:
            print("Product already exists!")
            return

        self.products[pid] = Product(pid, name, price, stock)
        print("Product added!")

    def show_products(self):
        print("\n--- PRODUCTS ---")
        for p in self.products.values():
            print(f"{p.pid} | {p.name} | Price: {p.price} | Stock: {p.stock}")

    # ---------------- CART SYSTEM ---------------- #

    def add_to_cart(self, pid, qty):
        product = self.products.get(pid)

        if not product:
            print("Product not found!")
            return

        if qty <= 0:
            print("Invalid quantity!")
            return

        if product.stock < qty:
            print("Not enough stock!")
            return

        self.cart[pid] = self.cart.get(pid, 0) + qty
        print(f"Added {qty} {product.name} to cart")

    def remove_from_cart(self, pid):
        if pid in self.cart:
            del self.cart[pid]
            print("Removed from cart")
        else:
            print("Item not in cart")

    def view_cart(self):
        print("\n--- CART ---")
        total = 0

        for pid, qty in self.cart.items():
            p = self.products[pid]
            cost = p.price * qty
            total += cost
            print(f"{p.name} x {qty} = {cost}")

        print(f"Total (before discount): {total}")
        return total

    # ---------------- CHECKOUT ---------------- #

    def checkout(self):
        if not self.cart:
            print("Cart is empty!")
            return

        total = self.view_cart()

        # Discount system
        discount = 0
        if total > 5000:
            discount = total * 0.10
        elif total > 2000:
            discount = total * 0.05

        final = total - discount

        print(f"Discount: {discount}")
        print(f"Final Amount: {final}")

        confirm = input("Confirm purchase? (yes/no): ")

        if confirm.lower() == "yes":
            for pid, qty in self.cart.items():
                self.products[pid].stock -= qty

            self.cart.clear()
            print("Purchase successful!")
        else:
            print("Order cancelled")

    # ---------------- SAVE/LOAD ---------------- #

    def save_data(self):
        data = {}
        for pid, p in self.products.items():
            data[pid] = p.to_dict()

        with open(self.file_name, "w") as f:
            json.dump(data, f, indent=4)

        print("Data saved!")

    def load_data(self):
        if not os.path.exists(self.file_name):
            return

        with open(self.file_name, "r") as f:
            data = json.load(f)

        for pid, info in data.items():
            self.products[pid] = Product(
                info["pid"],
                info["name"],
                info["price"],
                info["stock"]
            )


# ---------------- MAIN MENU ---------------- #

def main():
    shop = Shop()

    while True:
        print("\n===== SHOP SYSTEM =====")
        print("1. Add Product (Admin)")
        print("2. Show Products")
        print("3. Add to Cart")
        print("4. Remove from Cart")
        print("5. View Cart")
        print("6. Checkout")
        print("7. Save Data")
        print("8. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            pid = input("Product ID: ")
            name = input("Name: ")
            price = float(input("Price: "))
            stock = int(input("Stock: "))
            shop.add_product(pid, name, price, stock)

        elif choice == "2":
            shop.show_products()

        elif choice == "3":
            pid = input("Product ID: ")
            qty = int(input("Quantity: "))
            shop.add_to_cart(pid, qty)

        elif choice == "4":
            pid = input("Product ID: ")
            shop.remove_from_cart(pid)

        elif choice == "5":
            shop.view_cart()

        elif choice == "6":
            shop.checkout()

        elif choice == "7":
            shop.save_data()

        elif choice == "8":
            shop.save_data()
            print("Exiting shop system...")
            break

        else:
            print("Invalid choice!")


if __name__ == "__main__":
    main()
```

```python
import random

teams = {
    "Argentina": 96,
    "France": 94,
    "Spain": 93,
    "Brazil": 92,
    "England": 91,
    "Portugal": 90,
    "Germany": 89,
    "Netherlands": 88,
    "Italy": 87,
    "Belgium": 86,
    "Croatia": 85,
    "Uruguay": 84,
    "Morocco": 83,
    "Japan": 82,
    "USA": 81,
    "Mexico": 80
}


class Match:

    def __init__(self, team1, team2):
        self.team1 = team1
        self.team2 = team2

    def play(self):

        rating1 = teams[self.team1]
        rating2 = teams[self.team2]

        chance1 = rating1 / (rating1 + rating2)

        if random.random() < chance1:
            return self.team1

        return self.team2


class Tournament:

    def __init__(self, teams_list):
        self.teams = teams_list

    def next_round(self, current_teams):

        winners = []

        for i in range(0, len(current_teams), 2):

            match = Match(
                current_teams[i],
                current_teams[i + 1]
            )

            winner = match.play()

            print(
                f"{current_teams[i]} vs "
                f"{current_teams[i+1]}"
            )

            print(f"Winner ➜ {winner}\n")

            winners.append(winner)

        return winners

    def start(self):

        round_teams = self.teams

        stages = [
            "Round of 16",
            "Quarter Final",
            "Semi Final",
            "Final"
        ]

        for stage in stages:

            print("\n" + "=" * 35)

            print(stage)

            print("=" * 35)

            round_teams = self.next_round(
                round_teams
            )

        champion = round_teams[0]

        print("\n🏆 WORLD CUP CHAMPION")

        print(f"\n⭐ {champion} ⭐")


def main():

    tournament_teams = list(teams.keys())

    random.shuffle(
        tournament_teams
    )

    world_cup = Tournament(
        tournament_teams
    )

    world_cup.start()


if __name__ == "__main__":

    main()
```

```python
import json
import os
import hashlib
import secrets
import getpass
from base64 import b64encode, b64decode


class PasswordManager:

    def __init__(self, filename="vault.json"):
        self.filename = filename
        self.data = self.load_data()

    def derive_key(self, master_password, salt):
        return hashlib.pbkdf2_hmac(
            "sha256",
            master_password.encode(),
            salt,
            100000
        )

    def encrypt(self, text, key):

        text_bytes = text.encode()

        encrypted = bytearray()

        for i, byte in enumerate(text_bytes):
            encrypted.append(
                byte ^ key[i % len(key)]
            )

        return b64encode(
            encrypted
        ).decode()

    def decrypt(self, encrypted_text, key):

        encrypted_bytes = b64decode(
            encrypted_text
        )

        decrypted = bytearray()

        for i, byte in enumerate(
            encrypted_bytes
        ):

            decrypted.append(
                byte ^ key[i % len(key)]
            )

        return decrypted.decode()

    def load_data(self):

        if not os.path.exists(
            self.filename
        ):
            return {}

        with open(
            self.filename,
            "r"
        ) as file:

            return json.load(file)

    def save_data(self):

        with open(
            self.filename,
            "w"
        ) as file:

            json.dump(
                self.data,
                file,
                indent=4
            )

    def add_account(
        self,
        website,
        username,
        password,
        key
    ):

        encrypted_password = (
            self.encrypt(
                password,
                key
            )
        )

        self.data[website] = {
            "username": username,
            "password": encrypted_password
        }

        self.save_data()

        print(
            "Account saved successfully."
        )

    def view_account(
        self,
        website,
        key
    ):

        account = self.data.get(
            website
        )

        if not account:

            print(
                "Website not found."
            )

            return

        password = self.decrypt(
            account["password"],
            key
        )

        print("\nWebsite:", website)

        print(
            "Username:",
            account["username"]
        )

        print(
            "Password:",
            password
        )

    def delete_account(
        self,
        website
    ):

        if website in self.data:

            del self.data[website]

            self.save_data()

            print(
                "Deleted successfully."
            )

        else:

            print(
                "Website not found."
            )

    def list_accounts(
        self
    ):

        print("\nStored Accounts:\n")

        for site in self.data:

            print(site)


def main():

    manager = PasswordManager()

    print(
        "\n===== PASSWORD MANAGER ====="
    )

    master_password = getpass.getpass(
        "Master Password: "
    )

    salt = b"python_security"

    key = manager.derive_key(
        master_password,
        salt
    )

    while True:

        print(
            "\n1. Add Account"
        )

        print(
            "2. View Account"
        )

        print(
            "3. Delete Account"
        )

        print(
            "4. List Accounts"
        )

        print(
            "5. Exit"
        )

        choice = input(
            "Choice: "
        )

        if choice == "1":

            website = input(
                "Website: "
            )

            username = input(
                "Username: "
            )

            password = getpass.getpass(
                "Password: "
            )

            manager.add_account(
                website,
                username,
                password,
                key
            )

        elif choice == "2":

            website = input(
                "Website: "
            )

            manager.view_account(
                website,
                key
            )

        elif choice == "3":

            website = input(
                "Website: "
            )

            manager.delete_account(
                website
            )

        elif choice == "4":

            manager.list_accounts()

        elif choice == "5":

            print(
                "Goodbye."
            )

            break

        else:

            print(
                "Invalid choice."
            )


if __name__ == "__main__":

    main()
```

```python
import json
from datetime import datetime


class ExpenseTracker:

    def __init__(self):
        self.file_name = "expenses.json"
        self.expenses = []
        self.load_data()

    def load_data(self):

        try:
            with open(
                self.file_name,
                "r"
            ) as file:

                self.expenses = json.load(
                    file
                )

        except FileNotFoundError:

            self.expenses = []

    def save_data(self):

        with open(
            self.file_name,
            "w"
        ) as file:

            json.dump(
                self.expenses,
                file,
                indent=4
            )

    def add_expense(
        self,
        category,
        amount,
        description
    ):

        expense = {

            "category": category,

            "amount": amount,

            "description": description,

            "date": datetime.now().strftime(
                "%Y-%m-%d"
            )
        }

        self.expenses.append(
            expense
        )

        self.save_data()

        print(
            "Expense added successfully."
        )

    def view_expenses(self):

        if not self.expenses:

            print(
                "No expenses found."
            )

            return

        print(
            "\n===== EXPENSES =====\n"
        )

        for expense in self.expenses:

            print(
                f"{expense['date']} | "
                f"{expense['category']} | "
                f"{expense['amount']} | "
                f"{expense['description']}"
            )

    def total_expense(self):

        total = sum(

            expense["amount"]

            for expense in self.expenses
        )

        print(
            f"\nTotal Expense: {total}"
        )

    def category_report(self):

        report = {}

        for expense in self.expenses:

            category = expense[
                "category"
            ]

            report[category] = report.get(
                category,
                0
            ) + expense["amount"]

        print(
            "\n===== CATEGORY REPORT ====="
        )

        for key, value in report.items():

            print(
                f"{key}: {value}"
            )

    def search(self, keyword):

        found = False

        print(
            "\n===== SEARCH RESULT ====="
        )

        for expense in self.expenses:

            if keyword.lower() in expense[
                "description"
            ].lower():

                found = True

                print(
                    expense
                )

        if not found:

            print(
                "Nothing found."
            )


def main():

    tracker = ExpenseTracker()

    while True:

        print(
            "\n===== EXPENSE TRACKER ====="
        )

        print("1. Add Expense")

        print("2. View Expenses")

        print("3. Total Expense")

        print("4. Category Report")

        print("5. Search")

        print("6. Exit")

        choice = input(
            "Enter choice: "
        )

        if choice == "1":

            category = input(
                "Category: "
            )

            amount = float(
                input(
                    "Amount: "
                )
            )

            description = input(
                "Description: "
            )

            tracker.add_expense(
                category,
                amount,
                description
            )

        elif choice == "2":

            tracker.view_expenses()

        elif choice == "3":

            tracker.total_expense()

        elif choice == "4":

            tracker.category_report()

        elif choice == "5":

            keyword = input(
                "Keyword: "
            )

            tracker.search(
                keyword
            )

        elif choice == "6":

            print(
                "Goodbye."
            )

            break

        else:

            print(
                "Invalid option."
            )


if __name__ == "__main__":

    main()
```

```python
import random
import copy


class Sudoku:

    def __init__(self):

        self.board = [

            [0, 0, 0, 2, 6, 0, 7, 0, 1],

            [6, 8, 0, 0, 7, 0, 0, 9, 0],

            [1, 9, 0, 0, 0, 4, 5, 0, 0],

            [8, 2, 0, 1, 0, 0, 0, 4, 0],

            [0, 0, 4, 6, 0, 2, 9, 0, 0],

            [0, 5, 0, 0, 0, 3, 0, 2, 8],

            [0, 0, 9, 3, 0, 0, 0, 7, 4],

            [0, 4, 0, 0, 5, 0, 0, 3, 6],

            [7, 0, 3, 0, 1, 8, 0, 0, 0]
        ]

    def display(self):

        for row in range(9):

            if row % 3 == 0 and row != 0:

                print("-" * 25)

            for col in range(9):

                if col % 3 == 0 and col != 0:

                    print("|", end=" ")

                print(
                    self.board[row][col],
                    end=" "
                )

            print()

    def find_empty(self):

        for row in range(9):

            for col in range(9):

                if self.board[row][col] == 0:

                    return row, col

        return None

    def is_valid(
        self,
        number,
        position
    ):

        row, col = position

        for i in range(9):

            if (
                self.board[row][i]
                == number
                and col != i
            ):

                return False

        for i in range(9):

            if (
                self.board[i][col]
                == number
                and row != i
            ):

                return False

        box_x = col // 3

        box_y = row // 3

        for i in range(
            box_y * 3,
            box_y * 3 + 3
        ):

            for j in range(
                box_x * 3,
                box_x * 3 + 3
            ):

                if (
                    self.board[i][j]
                    == number
                    and (i, j) != position
                ):

                    return False

        return True

    def solve(self):

        empty = self.find_empty()

        if not empty:

            return True

        row, col = empty

        for number in range(1, 10):

            if self.is_valid(
                number,
                (row, col)
            ):

                self.board[row][col] = number

                if self.solve():

                    return True

                self.board[row][col] = 0

        return False

    def generate_puzzle(
        self,
        remove_cells=35
    ):

        solved = copy.deepcopy(
            self.board
        )

        self.solve()

        solved = copy.deepcopy(
            self.board
        )

        for _ in range(remove_cells):

            row = random.randint(
                0,
                8
            )

            col = random.randint(
                0,
                8
            )

            solved[row][col] = 0

        self.board = solved


def main():

    sudoku = Sudoku()

    print(
        "\n===== ORIGINAL PUZZLE =====\n"
    )

    sudoku.display()

    print(
        "\n===== SOLVING =====\n"
    )

    sudoku.solve()

    sudoku.display()

    print(
        "\n===== GENERATED PUZZLE =====\n"
    )

    sudoku.generate_puzzle()

    sudoku.display()


if __name__ == "__main__":

    main()
```

```python
import asyncio
import aiohttp
from bs4 import BeautifulSoup

urls = [
    "https://www.python.org",
    "https://github.com",
    "https://stackoverflow.com",
    "https://www.wikipedia.org",
]

async def fetch(session, url):
    try:
        async with session.get(url, timeout=10) as response:
            html = await response.text()

            soup = BeautifulSoup(html, "html.parser")
            title = soup.title.string.strip() if soup.title else "No Title"

            return {
                "url": url,
                "status": response.status,
                "title": title,
            }

    except Exception as e:
        return {
            "url": url,
            "error": str(e),
        }

async def main():
    async with aiohttp.ClientSession() as session:
        tasks = [fetch(session, url) for url in urls]
        results = await asyncio.gather(*tasks)

        print("=" * 70)
        for result in results:
            if "error" in result:
                print(f"❌ {result['url']}")
                print(f"Error : {result['error']}\n")
            else:
                print(f"🌐 URL    : {result['url']}")
                print(f"Status   : {result['status']}")
                print(f"Title    : {result['title']}")
                print("-" * 70)

if __name__ == "__main__":
    asyncio.run(main())
```

```python
import hashlib
import json
import time

class Block:
    def __init__(self, index, previous_hash, data):
        self.index = index
        self.timestamp = time.time()
        self.previous_hash = previous_hash
        self.data = data
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        block = json.dumps({
            "index": self.index,
            "timestamp": self.timestamp,
            "previous_hash": self.previous_hash,
            "data": self.data
        }, sort_keys=True)

        return hashlib.sha256(block.encode()).hexdigest()

class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]

    def create_genesis_block(self):
        return Block(0, "0", "Genesis Block")

    def latest_block(self):
        return self.chain[-1]

    def add_block(self, data):
        previous = self.latest_block()

        new_block = Block(
            previous.index + 1,
            previous.hash,
            data
        )

        self.chain.append(new_block)

    def display(self):
        for block in self.chain:
            print("="*60)
            print("Index :", block.index)
            print("Data :", block.data)
            print("Hash :", block.hash)
            print("Previous :", block.previous_hash)

blockchain = Blockchain()

blockchain.add_block("Alice pays Bob 10 BTC")
blockchain.add_block("Bob pays Charlie 5 BTC")
blockchain.add_block("Charlie pays David 2 BTC")

blockchain.display()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None


class BinarySearchTree:
    def __init__(self):
        self.root = None

    def insert(self, value):
        self.root = self._insert(self.root, value)

    def _insert(self, node, value):
        if node is None:
            return Node(value)

        if value < node.value:
            node.left = self._insert(node.left, value)
        elif value > node.value:
            node.right = self._insert(node.right, value)

        return node

    def search(self, value):
        return self._search(self.root, value)

    def _search(self, node, value):
        if node is None:
            return False

        if node.value == value:
            return True

        if value < node.value:
            return self._search(node.left, value)

        return self._search(node.right, value)

    def inorder(self):
        self._inorder(self.root)
        print()

    def _inorder(self, node):
        if node:
            self._inorder(node.left)
            print(node.value, end=" ")
            self._inorder(node.right)

    def preorder(self):
        self._preorder(self.root)
        print()

    def _preorder(self, node):
        if node:
            print(node.value, end=" ")
            self._preorder(node.left)
            self._preorder(node.right)

    def postorder(self):
        self._postorder(self.root)
        print()

    def _postorder(self, node):
        if node:
            self._postorder(node.left)
            self._postorder(node.right)
            print(node.value, end=" ")


# Driver Code
bst = BinarySearchTree()

numbers = [50, 30, 70, 20, 40, 60, 80]

for num in numbers:
    bst.insert(num)

print("Inorder Traversal:")
bst.inorder()

print("Preorder Traversal:")
bst.preorder()

print("Postorder Traversal:")
bst.postorder()

key = 60
if bst.search(key):
    print(f"{key} found in the tree.")
else:
    print(f"{key} not found.")
```

```python
import threading
import queue
import random
import time

buffer = queue.Queue(maxsize=5)

def producer():
    for _ in range(10):
        item = random.randint(1, 100)

        buffer.put(item)
        print(f"Produced: {item}")

        time.sleep(random.uniform(0.5, 1.5))

def consumer():
    for _ in range(10):
        item = buffer.get()

        print(f"Consumed: {item}")

        buffer.task_done()

        time.sleep(random.uniform(0.5, 2))

producer_thread = threading.Thread(target=producer)
consumer_thread = threading.Thread(target=consumer)

producer_thread.start()
consumer_thread.start()

producer_thread.join()
consumer_thread.join()

print("\nAll tasks completed successfully!")
```

```python
import heapq

graph = {
    'A': {'B': 4, 'C': 2},
    'B': {'A': 4, 'C': 1, 'D': 5},
    'C': {'A': 2, 'B': 1, 'D': 8, 'E': 10},
    'D': {'B': 5, 'C': 8, 'E': 2},
    'E': {'C': 10, 'D': 2}
}

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    pq = [(0, start)]

    while pq:
        current_distance, current_node = heapq.heappop(pq)

        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(pq, (distance, neighbor))

    return distances

print(dijkstra(graph, "A"))
```

```python
import threading
import time
import logging
from functools import wraps

# Configure logging
logging.basicConfig(
    level=logging.INFO,
    format="%(threadName)s - %(message)s"
)

# Decorator to measure execution time
def timer(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        logging.info(f"{func.__name__} finished in {end - start:.2f} seconds")
        return result
    return wrapper


class TaskExecutor:

    def __init__(self):
        self.results = []

    @timer
    def run_task(self, task_id):
        logging.info(f"Task {task_id} started")
        time.sleep(2)
        result = task_id ** 2
        self.results.append(result)
        logging.info(f"Task {task_id} completed → Result = {result}")

    def execute(self, total_tasks):
        threads = []

        for i in range(1, total_tasks + 1):
            thread = threading.Thread(
                target=self.run_task,
                args=(i,),
                name=f"Worker-{i}"
            )
            threads.append(thread)
            thread.start()

        for thread in threads:
            thread.join()

        logging.info(f"\nFinal Results: {self.results}")


if __name__ == "__main__":
    executor = TaskExecutor()
    executor.execute(5)
```

```python
from cryptography.fernet import Fernet
import os

class FileEncryptor:
    def __init__(self, key_file="secret.key"):
        self.key_file = key_file

        if os.path.exists(self.key_file):
            self.key = self.load_key()
        else:
            self.key = self.generate_key()

        self.cipher = Fernet(self.key)

    def generate_key(self):
        key = Fernet.generate_key()
        with open(self.key_file, "wb") as f:
            f.write(key)
        print("Encryption key created.")
        return key

    def load_key(self):
        with open(self.key_file, "rb") as f:
            return f.read()

    def encrypt_file(self, filename):
        with open(filename, "rb") as file:
            data = file.read()

        encrypted = self.cipher.encrypt(data)

        with open(filename + ".enc", "wb") as file:
            file.write(encrypted)

        print(f"{filename} encrypted successfully!")

    def decrypt_file(self, encrypted_file):
        with open(encrypted_file, "rb") as file:
            encrypted_data = file.read()

        decrypted = self.cipher.decrypt(encrypted_data)

        output_file = encrypted_file.replace(".enc", "_decrypted")

        with open(output_file, "wb") as file:
            file.write(decrypted)

        print(f"{encrypted_file} decrypted successfully!")
        print(f"Saved as: {output_file}")


if __name__ == "__main__":
    tool = FileEncryptor()

    while True:
        print("\n===== File Encryption Tool =====")
        print("1. Encrypt File")
        print("2. Decrypt File")
        print("3. Exit")

        choice = input("Choose an option: ")

        if choice == "1":
            file_name = input("Enter file name: ")
            if os.path.exists(file_name):
                tool.encrypt_file(file_name)
            else:
                print("File not found!")

        elif choice == "2":
            file_name = input("Enter encrypted file (.enc): ")
            if os.path.exists(file_name):
                tool.decrypt_file(file_name)
            else:
                print("File not found!")

        elif choice == "3":
            print("Goodbye!")
            break

        else:
            print("Invalid choice!")
```

```python
import sqlite3
import secrets
import string
import hashlib

class PasswordManager:

    def __init__(self):
        self.conn = sqlite3.connect("passwords.db")
        self.cursor = self.conn.cursor()

        self.cursor.execute("""
        CREATE TABLE IF NOT EXISTS passwords(
            id INTEGER PRIMARY KEY AUTOINCREMENT,
            website TEXT,
            username TEXT,
            password_hash TEXT
        )
        """)
        self.conn.commit()

    def generate_password(self, length=16):
        characters = string.ascii_letters + string.digits + "!@#$%^&*()"
        return ''.join(secrets.choice(characters) for _ in range(length))

    def hash_password(self, password):
        return hashlib.sha256(password.encode()).hexdigest()

    def save_password(self, website, username, password):
        password_hash = self.hash_password(password)

        self.cursor.execute("""
        INSERT INTO passwords(website, username, password_hash)
        VALUES (?, ?, ?)
        """, (website, username, password_hash))

        self.conn.commit()
        print("\nPassword Saved Successfully!")

    def show_passwords(self):
        self.cursor.execute("SELECT * FROM passwords")
        rows = self.cursor.fetchall()

        print("\n===== Saved Accounts =====")

        for row in rows:
            print(f"""
ID       : {row[0]}
Website  : {row[1]}
Username : {row[2]}
Hash     : {row[3]}
""")

manager = PasswordManager()

while True:

    print("""
====== PASSWORD MANAGER ======

1.Generate Password
2.Save Password
3.Show Saved Passwords
4.Exit
""")

    choice = input("Choose: ")

    if choice == "1":

        pwd = manager.generate_password()
        print("\nGenerated Password:", pwd)

    elif choice == "2":

        website = input("Website: ")
        username = input("Username: ")
        password = input("Password: ")

        manager.save_password(
            website,
            username,
            password
        )

    elif choice == "3":

        manager.show_passwords()

    elif choice == "4":

        break

    else:

        print("Invalid Option")
```

```python
import socket
import threading
from queue import Queue
import time

target = input("Enter Target IP or Domain: ")
start_port = int(input("Start Port: "))
end_port = int(input("End Port: "))

queue = Queue()
open_ports = []

print_lock = threading.Lock()


def scan_port(port):
    try:
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        sock.settimeout(0.5)

        result = sock.connect_ex((target, port))

        if result == 0:
            with print_lock:
                print(f"[OPEN] Port {port}")
                open_ports.append(port)

        sock.close()

    except:
        pass


def worker():
    while not queue.empty():
        port = queue.get()
        scan_port(port)
        queue.task_done()


for port in range(start_port, end_port + 1):
    queue.put(port)

start = time.time()

threads = []

for _ in range(100):  # 100 Worker Threads
    thread = threading.Thread(target=worker)
    thread.daemon = True
    thread.start()
    threads.append(thread)

queue.join()

end = time.time()

print("\n========== Scan Complete ==========")
print(f"Target      : {target}")
print(f"Open Ports  : {open_ports}")
print(f"Total Found : {len(open_ports)}")
print(f"Time Taken  : {end-start:.2f} seconds")
```

```python
class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.prev = None
        self.next = None


class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = {}

        self.head = Node(0, 0)
        self.tail = Node(0, 0)

        self.head.next = self.tail
        self.tail.prev = self.head

    def _add(self, node):
        node.prev = self.head
        node.next = self.head.next

        self.head.next.prev = node
        self.head.next = node

    def _remove(self, node):
        prev = node.prev
        nxt = node.next

        prev.next = nxt
        nxt.prev = prev

    def get(self, key):
        if key not in self.cache:
            return -1

        node = self.cache[key]
        self._remove(node)
        self._add(node)

        return node.value

    def put(self, key, value):

        if key in self.cache:
            self._remove(self.cache[key])

        node = Node(key, value)

        self.cache[key] = node
        self._add(node)

        if len(self.cache) > self.capacity:

            lru = self.tail.prev

            self._remove(lru)
            del self.cache[lru.key]


cache = LRUCache(3)

cache.put(1, "Python")
cache.put(2, "Java")
cache.put(3, "C++")

print(cache.get(2))

cache.put(4, "Rust")

print(cache.get(1))
print(cache.get(3))
print(cache.get(4))
```

```python
from cryptography.fernet import Fernet
from pathlib import Path

class FileEncryptor:
    def __init__(self, key_file="secret.key"):
        self.key_file = Path(key_file)

        if self.key_file.exists():
            self.key = self.key_file.read_bytes()
        else:
            self.key = Fernet.generate_key()
            self.key_file.write_bytes(self.key)

        self.cipher = Fernet(self.key)

    def encrypt(self, filename):
        path = Path(filename)

        data = path.read_bytes()
        encrypted = self.cipher.encrypt(data)

        encrypted_file = path.with_suffix(path.suffix + ".enc")
        encrypted_file.write_bytes(encrypted)

        print(f"Encrypted: {encrypted_file}")

    def decrypt(self, filename):
        path = Path(filename)

        encrypted_data = path.read_bytes()
        decrypted = self.cipher.decrypt(encrypted_data)

        output_file = path.with_suffix("")
        output_file.write_bytes(decrypted)

        print(f"Decrypted: {output_file}")


if __name__ == "__main__":
    tool = FileEncryptor()

    while True:
        print("\n=== File Encryption Tool ===")
        print("1. Encrypt File")
        print("2. Decrypt File")
        print("3. Exit")

        choice = input("Choice: ")

        if choice == "1":
            file = input("Enter file path: ")
            tool.encrypt(file)

        elif choice == "2":
            file = input("Enter encrypted file path: ")
            tool.decrypt(file)

        elif choice == "3":
            break

        else:
            print("Invalid choice.")
```

```python
import hashlib
import json
import time

class Block:
    def __init__(self, index, transactions, timestamp, previous_hash):
        self.index = index
        self.transactions = transactions
        self.timestamp = timestamp
        self.previous_hash = previous_hash
        self.nonce = 0
        self.hash = self.calculate_hash()

    def calculate_hash(self):
        block = json.dumps({
            "index": self.index,
            "transactions": self.transactions,
            "timestamp": self.timestamp,
            "previous_hash": self.previous_hash,
            "nonce": self.nonce
        }, sort_keys=True).encode()

        return hashlib.sha256(block).hexdigest()

    def mine(self, difficulty):
        target = "0" * difficulty

        while self.hash[:difficulty] != target:
            self.nonce += 1
            self.hash = self.calculate_hash()

        print(f"Block {self.index} Mined!")
        print(f"Hash: {self.hash}\n")


class Blockchain:
    def __init__(self):
        self.chain = [self.create_genesis_block()]
        self.pending_transactions = []
        self.difficulty = 4

    def create_genesis_block(self):
        return Block(0, [], time.time(), "0")

    def get_latest_block(self):
        return self.chain[-1]

    def add_transaction(self, sender, receiver, amount):
        self.pending_transactions.append({
            "sender": sender,
            "receiver": receiver,
            "amount": amount
        })

    def mine_pending_transactions(self):
        block = Block(
            len(self.chain),
            self.pending_transactions,
            time.time(),
            self.get_latest_block().hash
        )

        block.mine(self.difficulty)

        self.chain.append(block)
        self.pending_transactions = []

    def is_chain_valid(self):
        for i in range(1, len(self.chain)):
            current = self.chain[i]
            previous = self.chain[i - 1]

            if current.hash != current.calculate_hash():
                return False

            if current.previous_hash != previous.hash:
                return False

        return True

    def show_chain(self):
        for block in self.chain:
            print("=" * 60)
            print(f"Block #{block.index}")
            print(f"Hash          : {block.hash}")
            print(f"Previous Hash : {block.previous_hash}")
            print(f"Nonce         : {block.nonce}")
            print(f"Transactions  : {block.transactions}")


if __name__ == "__main__":
    blockchain = Blockchain()

    blockchain.add_transaction("Alice", "Bob", 50)
    blockchain.add_transaction("Bob", "Charlie", 20)

    blockchain.mine_pending_transactions()

    blockchain.add_transaction("Charlie", "David", 10)

    blockchain.mine_pending_transactions()

    blockchain.show_chain()

    print("\nBlockchain Valid:", blockchain.is_chain_valid())
```

```python
from abc import ABC, abstractmethod

class PluginMeta(type):
    registry = {}

    def __new__(cls, name, bases, attrs):
        new_class = super().__new__(cls, name, bases, attrs)
        if name != "Plugin":
            cls.registry[name] = new_class
        return new_class


class Plugin(metaclass=PluginMeta):
    @abstractmethod
    def run(self, data):
        pass


class UpperCasePlugin(Plugin):
    def run(self, data):
        return data.upper()


class ReversePlugin(Plugin):
    def run(self, data):
        return data[::-1]


class WordCountPlugin(Plugin):
    def run(self, data):
        return len(data.split())


text = "Advanced Python Programming"

for plugin_name, plugin_class in PluginMeta.registry.items():
    plugin = plugin_class()
    print(f"{plugin_name}: {plugin.run(text)}")
```

```python
from typing import Type, Dict, Any


class Container:
    def __init__(self):
        self._services: Dict[Type, Any] = {}

    def register(self, interface: Type, implementation: Any):
        self._services[interface] = implementation

    def resolve(self, interface: Type):
        service = self._services.get(interface)
        if service is None:
            raise ValueError(f"{interface.__name__} not registered")

        return service() if isinstance(service, type) else service


class Database:
    def connect(self):
        return "Connected to PostgreSQL"


class UserService:
    def __init__(self, db):
        self.db = db

    def users(self):
        return self.db.connect()


container = Container()
container.register(Database, Database)

db = container.resolve(Database)
service = UserService(db)

print(service.users())
```

```python
import time
from functools import wraps


def retry(max_attempts=5, delay=1):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            current_delay = delay

            for attempt in range(max_attempts):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Attempt {attempt + 1} failed: {e}")

                    if attempt == max_attempts - 1:
                        raise

                    time.sleep(current_delay)
                    current_delay *= 2

        return wrapper
    return decorator


counter = 0


@retry()
def unstable():
    global counter
    counter += 1

    if counter < 4:
        raise RuntimeError("Temporary Error")

    return "Success"


print(unstable())
```

```python
import asyncio
import time


class RateLimiter:

    def __init__(self, limit, period):
        self.limit = limit
        self.period = period
        self.calls = []

    async def wait(self):

        while len(self.calls) >= self.limit:

            now = time.monotonic()

            self.calls = [
                t for t in self.calls
                if now - t < self.period
            ]

            await asyncio.sleep(0.1)

        self.calls.append(time.monotonic())


limiter = RateLimiter(3, 5)


async def request(i):
    await limiter.wait()
    print(f"Request {i}")


async def main():
    await asyncio.gather(*(request(i) for i in range(10)))


asyncio.run(main())
```

```python
import socket
from concurrent.futures import ThreadPoolExecutor

HOST = "127.0.0.1"
PORT = 8080


def handle_client(client):
    request = client.recv(4096).decode()

    print(request.splitlines()[0])

    response = (
        "HTTP/1.1 200 OK\r\n"
        "Content-Type: text/html\r\n\r\n"
        "<h1>Hello from Pure Python HTTP Server</h1>"
    )

    client.sendall(response.encode())
    client.close()


server = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
server.bind((HOST, PORT))
server.listen()

print(f"Listening on http://{HOST}:{PORT}")

with ThreadPoolExecutor(max_workers=20) as pool:
    while True:
        client, addr = server.accept()
        pool.submit(handle_client, client)
```

```python
import time
from functools import wraps

def retry(max_attempts=5, delay=1):
    def decorator(func):
        @wraps(func)
        def wrapper(*args, **kwargs):
            current_delay = delay
            for attempt in range(max_attempts):
                try:
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Attempt {attempt+1} failed: {e}")
                    if attempt == max_attempts - 1:
                        raise
                    time.sleep(current_delay)
                    current_delay *= 2
        return wrapper
    return decorator

@retry(max_attempts=3)
def unstable():
    import random
    if random.random() < 0.8:
        raise ValueError("Random failure")
    return "Success"

print(unstable())
```

```python
import asyncio

async def producer(queue):
    for i in range(10):
        await asyncio.sleep(0.5)
        await queue.put(i)
        print(f"Produced {i}")

async def consumer(queue):
    while True:
        item = await queue.get()
        print(f"Consumed {item}")
        queue.task_done()

async def main():
    queue = asyncio.Queue()

    consumer_task = asyncio.create_task(consumer(queue))
    await producer(queue)
    await queue.join()

    consumer_task.cancel()

asyncio.run(main())
```

```python
from collections import OrderedDict

class LRUCache:
    def __init__(self, capacity):
        self.capacity = capacity
        self.cache = OrderedDict()

    def get(self, key):
        if key not in self.cache:
            return -1
        self.cache.move_to_end(key)
        return self.cache[key]

    def put(self, key, value):
        self.cache[key] = value
        self.cache.move_to_end(key)

        if len(self.cache) > self.capacity:
            self.cache.popitem(last=False)

cache = LRUCache(3)

cache.put(1, "A")
cache.put(2, "B")
cache.put(3, "C")

print(cache.get(2))

cache.put(4, "D")

print(cache.cache)
```

```python
import time
from functools import wraps

def ttl_cache(seconds):
    def decorator(func):
        cache = {}

        @wraps(func)
        def wrapper(*args):
            now = time.time()

            if args in cache:
                result, timestamp = cache[args]
                if now - timestamp < seconds:
                    return result

            result = func(*args)
            cache[args] = (result, now)
            return result

        return wrapper
    return decorator


@ttl_cache(seconds=5)
def expensive(x):
    print("Calculating...")
    return x ** 2

print(expensive(10))
print(expensive(10))
```

```python
import functools

class LogMethods(type):

    def __new__(cls, name, bases, attrs):

        for key, value in attrs.items():

            if callable(value):

                @functools.wraps(value)
                def wrapper(*args, __func=value, **kwargs):
                    print("Calling", __func.__name__)
                    return __func(*args, **kwargs)

                attrs[key] = wrapper

        return super().__new__(cls, name, bases, attrs)

class Demo(metaclass=LogMethods):

    def hello(self):
        print("Hello!")

Demo().hello()
```

```python
import asyncio

class RateLimiter:

    def __init__(self, rate):
        self.semaphore = asyncio.Semaphore(rate)

    async def run(self, coro):

        async with self.semaphore:
            return await coro

async def worker(i):
    await asyncio.sleep(1)
    print(i)

async def main():

    limiter = RateLimiter(3)

    await asyncio.gather(
        *(limiter.run(worker(i)) for i in range(10))
    )

asyncio.run(main())
```

```python
from functools import wraps

def memoize(func):
    cache = {}

    @wraps(func)
    def wrapper(*args):
        if args not in cache:
            cache[args] = func(*args)
        return cache[args]

    return wrapper


@memoize
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n - 1) + fibonacci(n - 2)


print(fibonacci(100))
```

```python
def solve(board):
    for row in range(9):
        for col in range(9):
            if board[row][col] == 0:

                for num in range(1,10):
                    if valid(board,row,col,num):
                        board[row][col]=num

                        if solve(board):
                            return True

                        board[row][col]=0
                return False
    return True

def valid(board,row,col,num):
    if num in board[row]:
        return False

    for i in range(9):
        if board[i][col]==num:
            return False

    sr=row-row%3
    sc=col-col%3

    for i in range(sr,sr+3):
        for j in range(sc,sc+3):
            if board[i][j]==num:
                return False

    return True
```
