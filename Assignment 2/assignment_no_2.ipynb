# Assignment 1: Email Validator (String Filtering + Loops)

```python
email = input("Enter your email address: ")

valid = True
special = "!#$%^&*()-_=+[]{}|;:'\",<>/?`~"

if "@" not in email or email.count("@") != 1:
    valid = False

if not (email.endswith(".com") or email.endswith(".org")):
    valid = False

if " " in email or ".." in email:
    valid = False

if valid:
    user, domain = email.split("@")

    if len(user) == 0:
        valid = False
    elif user[0] in special or user[-1] in special:
        valid = False

    if len(domain) == 0:
        valid = False
    elif domain[0] in special:
        valid = False

print("Valid Email" if valid else "Invalid Email")
```

# Assignment 2: Count Word Frequency in Customer Reviews 

```python
text_str = input("Enter a review or comment: ")
text_str = text_str.lower()

words = text_str.split()

print("{", end="")

for i in range(len(words)):

    if words[i] in words[:i]:
        continue

    count = words.count(words[i])

    last_unique = True
    for j in range(i+1, len(words)):
        if words[j] not in words[:j]:
            last_unique = False
            break

    if last_unique:
        print(f"'{words[i]}': {count}", end="")
    else:
        print(f"'{words[i]}': {count}", end=", ")

print("}")
```

# Assignment 3: Password Strength Checker

```python
special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>/?`~"

valid_pass = False

while not valid_pass:

    pwd = input("Enter a password: ")

    upper = lower = digit = special = False

    for ch in pwd:
        if ch.isupper():
            upper = True
        elif ch.islower():
            lower = True
        elif ch.isdigit():
            digit = True
        elif ch in special_characters:
            special = True

    length_ok = len(pwd) >= 8

    if upper and lower and digit and special and length_ok:
        print("Strong Password")
        valid_pass = True
    else:
        print("Weak Password")

        if not upper:
            print("- Add at least one uppercase letter (A-Z)")
        if not lower:
            print("- Add at least one lowercase letter (a-z)")
        if not digit:
            print("- Add at least one number (0-9)")
        if not special:
            print("- Add at least one special character (!,@,#,$,%, etc.)")
        if not length_ok:
            print("- Password must be at least 8 characters long")
```

# Assignment 4: Username Generator

```python
import random

full_name = input("Enter full name: ")

parts = []
temp = ""

for ch in full_name:
    if ch != " ":
        temp += ch
    else:
        if temp != "":
            parts.append(temp)
            temp = ""

if temp != "":
    parts.append(temp)

first_name = parts[0]
last_name = parts[-1]

first_part = ""
i = 0
while i < 3 and i < len(first_name):
    first_part += first_name[i]
    i += 1

last_part = ""
j = len(last_name) - 3
if j < 0:
    j = 0

while j < len(last_name):
    last_part += last_name[j]
    j += 1

num = random.randint(10, 99)

username = first_part.lower() + last_part.lower() + str(num)

print("Generated Username:", username)
```

# Assignment 5: SMS Word Counter

```python
sms = input("Enter a demo message: ")

char_count = 0
word_count = 0
inside_word = False

for ch in sms:

    char_count += 1  

    if ch != " " and inside_word == False:
        word_count += 1
        inside_word = True
    elif ch == " ":
        inside_word = False

if char_count > 160:
    status = "Exceed Limit"
else:
    status = "Within Limit"

print("Words =", word_count, ", Characters =", char_count, ", Status =", status)
```

# Assignment 6: Remove Duplicates from Contact List

```python
contact_str = "Rahim, Karim, Rahim, Rahim, Sultana"

names = []
temp = ""

for ch in contact_str:
    if ch == ",":
        cleaned = temp.strip()
        if cleaned != "":
            names.append(cleaned)
        temp = ""
    else:
        temp += ch

last_clean = temp.strip()
if last_clean != "":
    names.append(last_clean)

unique_contacts = []

for nm in names:
    seen = False

    for u in unique_contacts:
        if u == nm:
            seen = True
            break

    if not seen:
        unique_contacts.append(nm)

result = ""
idx = 0

while idx < len(unique_contacts):
    result += unique_contacts[idx]
    if idx != len(unique_contacts) - 1:
        result += ", "
    idx += 1

print('Output: "' + result + '"')
```

# Assignment 7: Detect Suspicious Words in Text (Moderation) 

```python
banned_words = ['spam', 'scam', 'fake']

text = input("Enter content: ").lower()

words = []
temp = ""

for ch in text:
    if ch != " ":
        temp += ch
    else:
        if temp != "":
            words.append(temp)
            temp = ""

if temp != "":
    words.append(temp)

alert = False
i = 0

while i < len(words) and not alert:
    j = 0
    while j < len(banned_words):
        if words[i] == banned_words[j]:
            alert = True
            break
        j += 1
    i += 1

if alert:
    print("Warning: Suspicious content detected.")
else:
    print("Safe: No banned words found")
```

# Assignment 8: Invoice Code Formatter

```python
num = int(input("Enter an invoice number: "))
num_str = str(num)

total_digits = 5
pad_zeros = total_digits - len(num_str)

zeros = ""
i = 0
while i < pad_zeros:
    zeros += "0"
    i += 1

invoice_code = "INV" + zeros + num_str
print("Generated Code:", invoice_code)
```

# Assignment 9: Count Hashtags and Mentions (Social Media Analytics)

```python
post_text = input("Enter your post: ")

hashtags = 0
mentions = 0
i = 0

while i < len(post_text):
    char = post_text[i]
    if char == "#":
        hashtags += 1
    elif char == "@":
        mentions += 1
    i += 1

print("Hashtags =", hashtags, ", Mentions =", mentions)
```

# Assignment 10: Word Reversal for Data Encryption

```python
sentence = input("Enter a sentence: ")

word = ""
reversed_words = []

for ch in sentence:
    if ch != " ":
        word += ch
    else:
        if word != "":
            rev_word = ""
            i = len(word) - 1
            while i >= 0:
                rev_word += word[i]
                i -= 1
            reversed_words.append(rev_word)
            word = ""

if word != "":
    rev_word = ""
    i = len(word) - 1
    while i >= 0:
        rev_word += word[i]
        i -= 1
    reversed_words.append(rev_word)

result = ""
idx = 0
while idx < len(reversed_words):
    result += reversed_words[idx]
    if idx != len(reversed_words) - 1:
        result += " "
    idx += 1

print("Reversed Words Sentence:", result)
```
